```python
import pandas as pd

# Load the data (assuming you have both datasets loaded)
# excel_path = '/content/AR-1.xlsx'
# all_floor_data = pd.read_excel(excel_path, sheet_name='Program Table Input 2 - Floor')
# all_block_data = pd.read_excel(excel_path, sheet_name='your_block_sheet_name')

def initialize_floor_assignments(floor_df):
    """Initialize floor assignments with remaining capacity tracking"""
    assignments = {}
    for _, row in floor_df.iterrows():
        floor = row['Name'].strip()
        assignments[floor] = {
            'remaining_area': row['Usable_Area_(SQM)'],
            'remaining_capacity': row['Max_Assignable_Floor_loading_Capacity'],
            'assigned_blocks': [],
            'assigned_departments': set(),
            'total_circulation_area': 0,
            'total_capacity': 0
        }
    return assignments

def convert_sqft_to_sqm(sqft):
    """Convert square feet to square meters (1 sqft = 0.092903 sqm)"""
    return sqft * 0.092903

def get_destination_blocks(block_df):
    """Filter blocks that are destinations and group them by Destination_Group"""
    destination_blocks = block_df[block_df['Typical_Destination'] == 'Destination'].copy()

    # Group by Destination_Group
    destination_groups = {}
    for _, row in destination_blocks.iterrows():
        group = row['Destination_Group']
        if group not in destination_groups:
            destination_groups[group] = []
        destination_groups[group].append(row)

    return destination_groups

def calculate_group_requirements(group_blocks):
    """Calculate total area and capacity requirements for a group"""
    total_circulation_area_sqft = sum(block['Cumulative_Block_Circulation_Area_(SQFT)'] for block in group_blocks)
    total_circulation_area_sqm = convert_sqft_to_sqm(total_circulation_area_sqft)
    total_capacity = sum(block['Block_Capacity'] for block in group_blocks)

    return total_circulation_area_sqm, total_capacity

def can_fit_in_floor(floor_data, group_area_sqm, group_capacity):
    """Check if a group can fit in a floor based on constraints"""
    area_fits = floor_data['remaining_area'] >= group_area_sqm
    capacity_fits = floor_data['remaining_capacity'] >= group_capacity
    return area_fits and capacity_fits

def assign_destination_blocks_to_floors(floor_df, block_df):
    """
    Main function to assign destination blocks to floors
    """
    # Initialize floor assignments
    floor_assignments = initialize_floor_assignments(floor_df)

    # Get sorted list of floors (assuming first two floors are Ground Floor and First Floor)
    floor_names = list(floor_assignments.keys())
    target_floors = floor_names[:2]  # First two floors

    print(f"Target floors for destination blocks: {target_floors}")
    print("=" * 50)

    # Get destination block groups
    destination_groups = get_destination_blocks(block_df)

    print(f"Found {len(destination_groups)} destination groups:")
    for group_name, blocks in destination_groups.items():
        print(f"  - {group_name}: {len(blocks)} blocks")
    print("=" * 50)

    # Sort groups by total area requirement (largest first for better packing)
    group_requirements = {}
    for group_name, blocks in destination_groups.items():
        area_sqm, capacity = calculate_group_requirements(blocks)
        group_requirements[group_name] = {
            'area_sqm': area_sqm,
            'capacity': capacity,
            'blocks': blocks
        }

    # Sort by area requirement (descending)
    sorted_groups = sorted(group_requirements.items(),
                          key=lambda x: x[1]['area_sqm'],
                          reverse=True)

    # Assignment tracking
    successfully_assigned = []
    failed_assignments = []

    # Try to assign each group to floors
    for group_name, group_data in sorted_groups:
        assigned = False

        print(f"\nTrying to assign group '{group_name}':")
        print(f"  Required area: {group_data['area_sqm']:.2f} SQM")
        print(f"  Required capacity: {group_data['capacity']}")

        # Try each target floor in order
        for floor_name in target_floors:
            if can_fit_in_floor(floor_assignments[floor_name],
                              group_data['area_sqm'],
                              group_data['capacity']):

                # Assign the group to this floor
                floor_assignments[floor_name]['assigned_blocks'].extend(group_data['blocks'])
                floor_assignments[floor_name]['remaining_area'] -= group_data['area_sqm']
                floor_assignments[floor_name]['remaining_capacity'] -= group_data['capacity']
                floor_assignments[floor_name]['total_circulation_area'] += group_data['area_sqm']
                floor_assignments[floor_name]['total_capacity'] += group_data['capacity']
                floor_assignments[floor_name]['assigned_departments'].add(group_name)

                print(f"  ✓ Assigned to {floor_name}")
                print(f"    Remaining area: {floor_assignments[floor_name]['remaining_area']:.2f} SQM")
                print(f"    Remaining capacity: {floor_assignments[floor_name]['remaining_capacity']}")

                successfully_assigned.append((group_name, floor_name))
                assigned = True
                break
            else:
                remaining_area = floor_assignments[floor_name]['remaining_area']
                remaining_capacity = floor_assignments[floor_name]['remaining_capacity']
                print(f"  ✗ Cannot fit in {floor_name} (Available: {remaining_area:.2f} SQM, {remaining_capacity} capacity)")

        if not assigned:
            failed_assignments.append(group_name)
            print(f"  ✗ Could not assign group '{group_name}' to any target floor")

    return floor_assignments, successfully_assigned, failed_assignments

def print_assignment_summary(floor_assignments, successfully_assigned, failed_assignments, target_floors):
    """Print a summary of the floor assignments"""
    print("\n" + "=" * 60)
    print("FLOOR ASSIGNMENT SUMMARY")
    print("=" * 60)

    for floor_name in target_floors:
        floor_data = floor_assignments[floor_name]
        print(f"\n{floor_name.upper()}:")
        print(f"  Assigned Groups: {len(floor_data['assigned_departments'])}")
        print(f"  Assigned Blocks: {len(floor_data['assigned_blocks'])}")
        print(f"  Total Area Used: {floor_data['total_circulation_area']:.2f} SQM")
        print(f"  Total Capacity Used: {floor_data['total_capacity']}")
        print(f"  Remaining Area: {floor_data['remaining_area']:.2f} SQM")
        print(f"  Remaining Capacity: {floor_data['remaining_capacity']}")

        if floor_data['assigned_departments']:
            print(f"  Department Groups: {', '.join(floor_data['assigned_departments'])}")

    print(f"\nSUCCESSFULLY ASSIGNED: {len(successfully_assigned)} groups")
    for group_name, floor_name in successfully_assigned:
        print(f"  - {group_name} → {floor_name}")

    if failed_assignments:
        print(f"\nFAILED ASSIGNMENTS: {len(failed_assignments)} groups")
        for group_name in failed_assignments:
            print(f"  - {group_name}")

    print("=" * 60)

def create_assignment_dataframe(floor_assignments):
    """Create a DataFrame showing the assignment results"""
    assignment_data = []

    for floor_name, floor_data in floor_assignments.items():
        for block in floor_data['assigned_blocks']:
            assignment_data.append({
                'Floor': floor_name,
                'Block_ID': block['Block_ID'],
                'Block_Name': block['Block_Name'],
                'Block_Instance_Name': block['Block_Instance_Name'],
                'Destination_Group': block['Destination_Group'],
                'Block_Capacity': block['Block_Capacity'],
                'Circulation_Area_SQFT': block['Cumulative_Block_Circulation_Area_(SQFT)'],
                'Circulation_Area_SQM': convert_sqft_to_sqm(block['Cumulative_Block_Circulation_Area_(SQFT)'])
            })

    return pd.DataFrame(assignment_data)

# Example usage:
if __name__ == "__main__":
    # Assuming you have loaded your data
    # floor_assignments, successfully_assigned, failed_assignments = assign_destination_blocks_to_floors(all_floor_data, all_block_data)

    # Print summary
    # target_floors = list(floor_assignments.keys())[:2]
    # print_assignment_summary(floor_assignments, successfully_assigned, failed_assignments, target_floors)

    # Create assignment DataFrame
    # assignment_df = create_assignment_dataframe(floor_assignments)
    # print("\nAssignment DataFrame:")
    # print(assignment_df)

    print("Floor assignment algorithm ready!")
    print("To use:")
    print("1. Load your floor and block data")
    print("2. Call: floor_assignments, successfully_assigned, failed_assignments = assign_destination_blocks_to_floors(all_floor_data, all_block_data)")
    print("3. View results with the summary and DataFrame functions")
```

```python
import pandas as pd
from collections import defaultdict

# Load data (simulated placeholder)
# all_floor_data = pd.read_csv("floor_data.csv")
# all_block_data = pd.read_csv("block_data.csv")

# Conversion constant
SQFT_TO_SQM = 0.092903

# Global assignment dictionary
assignment = {}

def assign_destination_blocks(all_floor_data, all_block_data):
    global assignment

    # Filter blocks with Typical_Destination = Destination
    destination_blocks = all_block_data[all_block_data["Typical_Destination"].isin(["Destination", "Both"])]


    # Drop rows with missing Destination_Group or area
    destination_blocks = destination_blocks.dropna(subset=["Destination_Group", "Cumulative_Block_Circulation_Area_(SQFT)"])

    # Group blocks by Destination_Group
    group_map = defaultdict(list)
    for _, block in destination_blocks.iterrows():
        group_map[block["Destination_Group"]].append(block)

    # Prepare floors state
    floor_state = []
    for _, floor in all_floor_data.iterrows():
        floor_state.append({
            "Floor_Name": floor["Name"],
            "Usable_Area_(SQM)": floor["Usable_Area_(SQM)"],
            "Blocks": []
        })

    assignment.clear()

    # Assign groups to floors
    for group_id, blocks in group_map.items():
        total_group_area_sqm = sum(block["Cumulative_Block_Circulation_Area_(SQFT)"] * SQFT_TO_SQM for block in blocks)

        assigned = False
        for floor in floor_state:
            if floor["Usable_Area_(SQM)"] >= total_group_area_sqm:
                floor["Usable_Area_(SQM)"] -= total_group_area_sqm
                floor["Blocks"].extend([block["Block_Instance_Name"] for block in blocks])
                for block in blocks:
                    assignment[block["Block_Instance_Name"]] = floor["Floor_Name"]
                assigned = True
                break

        if not assigned:
            print(f"⚠ Could not assign group '{group_id}' with area {round(total_group_area_sqm, 2)} SQM")

    return floor_state
```

```python
import pandas as pd

# ----------------------------------------
# Step 1: Load Input Sheets (assumed already loaded)
# ----------------------------------------

excel_path = '/content/AR-1.xlsx'  # <-- adjust path if needed

# 1.1 Floors
all_floor_data = pd.read_excel(excel_path, sheet_name='Program Table Input 2 - Floor')
all_floor_data.columns = all_floor_data.columns.str.strip()

# 1.2 Blocks
all_block_data = pd.read_excel(excel_path, sheet_name='Program Table Input 1 - Block')
all_block_data.columns = all_block_data.columns.str.strip()

# 1.3 Department Split
department_split_data = pd.read_excel(excel_path, sheet_name='Department Split', skiprows=1)
department_split_data.columns = department_split_data.columns.str.strip()
department_split_data = department_split_data.rename(
    columns={'BU_Department_Sub-Department': 'Department_Sub-Department'}
)

# 1.4 Adjacency
xls = pd.ExcelFile(excel_path)
adjacency_sheet_name = [name for name in xls.sheet_names if "Adjacency" in name][0]
raw_data = xls.parse(adjacency_sheet_name, header=1, index_col=0)
adjacency_data = raw_data.apply(pd.to_numeric, errors='coerce')
adjacency_data.index   = adjacency_data.index.str.strip()
adjacency_data.columns = adjacency_data.columns.str.strip()

# 1.5 De-Centralized Logic
df_logic = pd.read_excel(excel_path, sheet_name='De-Centralized Logic', header=None)
De_Centralized_data = {}
current_section = None
for _, row in df_logic.iterrows():
    first_cell = str(row[0]).strip() if pd.notna(row[0]) else ""
    if first_cell in ["Centralised", "Semi Centralized", "DeCentralised"]:
        current_section = first_cell
        De_Centralized_data[current_section] = {}
    elif current_section and first_cell == "Destination":
        De_Centralized_data[current_section]["Destination"] = int(row[1]) if pd.notna(row[1]) else None

# ----------------------------------------
# Step 2: Preprocess block & department‐split
# ----------------------------------------

# 2.1 Convert cumulative area from SQFT to SQM
all_block_data['Cumulative_Area_SQM'] = all_block_data['Cumulative_Block_Circulation_Area_(SQFT)'] / 10.764

# 2.2 Split "Destination" vs. "Typical"
destination_blocks = all_block_data[all_block_data['Typical_Destination'] == 'Destination'].copy()
typical_blocks     = all_block_data[all_block_data['Typical_Destination'] == 'Typical'].copy()

# 2.3 Build dept_split_dict
dept_split_dict = department_split_data.set_index('Department_Sub-Department')['Splittable'].to_dict()

# 2.4 Function to initialize floor assignments
def initialize_floor_assignments(floor_df):
    assignments = {}
    for _, row in floor_df.iterrows():
        floor = row['Name'].strip()
        assignments[floor] = {
            'remaining_area': row['Usable_Area_(SQM)'],
            'remaining_capacity': row['Max_Assignable_Floor_loading_Capacity'],
            'assigned_blocks': [],       # Will store block‐row dicts
            'assigned_departments': set(),
            'ME_area': 0.0,
            'WE_area': 0.0,
            'US_area': 0.0,
            'Support_area': 0.0,
            'Speciality_area': 0.0
        }
    return assignments

floors = list(initialize_floor_assignments(all_floor_data).keys())

# ----------------------------------------
# Step 3: Core stacking function (store full block dict)
# ----------------------------------------

def run_stack_plan(mode):
    """
    mode: 'centralized' or 'decentralized'
    Returns three DataFrames:
      1) detailed_df      – block‐to‐floor assignments
      2) floor_summary_df – floor‐wise totals (blocks, area, occupancy)
      3) space_mix_df     – floor‐wise % breakdown by space‐mix
    """
    assignments = initialize_floor_assignments(all_floor_data)

    def floors_allowed_for_dest(_group_name):
        if mode == 'centralized':
            return 2
        elif mode == 'decentralized':
            return len(floors)
        else:
            return len(floors)

    # 3.1 Assign Destination Blocks
    for group_name, group_df in destination_blocks.groupby('Destination_Group'):
        sorted_blocks = group_df.sort_values(by='Cumulative_Area_SQM', ascending=False).to_dict('records')
        max_floors = floors_allowed_for_dest(group_name)
        target_floors = floors[:max_floors]

        for blk in sorted_blocks:
            blk_area     = blk['Cumulative_Area_SQM']
            blk_capacity = blk['Max_Occupancy_with_Capacity']
            blk_dept     = blk['Department_Sub_Department'].strip()
            placed       = False

            # 3.1.a Try allowed destination floors
            for fl in target_floors:
                if assignments[fl]['remaining_area'] >= blk_area:
                    assignments[fl]['assigned_blocks'].append(blk)
                    assignments[fl]['remaining_area'] -= blk_area
                    assignments[fl]['remaining_capacity'] -= blk_capacity
                    assignments[fl]['assigned_departments'].add(blk_dept)
                    placed = True
                    break

            # 3.1.b If not placed, move up
            if not placed:
                for fl in floors[max_floors:]:
                    if assignments[fl]['remaining_area'] >= blk_area:
                        assignments[fl]['assigned_blocks'].append(blk)
                        assignments[fl]['remaining_area'] -= blk_area
                        assignments[fl]['remaining_capacity'] -= blk_capacity
                        assignments[fl]['assigned_departments'].add(blk_dept)
                        placed = True
                        break

            if not placed:
                print(f"Warning: Could not place destination block '{blk['Block_Name']}'")

    # 3.2 Assign Typical Blocks
    typical_records = typical_blocks.sort_values(by='Cumulative_Area_SQM', ascending=False).to_dict('records')

    for blk in typical_records:
        blk_area        = blk['Cumulative_Area_SQM']
        blk_capacity    = blk['Max_Occupancy_with_Capacity']
        blk_dept        = blk['Department_Sub_Department'].strip()
        blk_spacemix    = blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip()
        splittable_flag = dept_split_dict.get(blk_dept, 0)

        # 3.2.a Candidate floors with enough area
        candidates = [fl for fl in floors if assignments[fl]['remaining_area'] >= blk_area]
        assigned_floor = None

        # 3.2.b If non‐splittable, prefer floor already containing that dept
        if splittable_flag == 1:
            for fl in candidates:
                if blk_dept in assignments[fl]['assigned_departments']:
                    assigned_floor = fl
                    break

        # 3.2.c Otherwise, compute adjacency_score ×1000 + leftover area
        if assigned_floor is None and candidates:
            best_score = None
            best_floor = None
            for fl in candidates:
                existing_depts = assignments[fl]['assigned_departments']
                if blk_dept in adjacency_data.index:
                    row = adjacency_data.loc[blk_dept]
                    adj_score = sum(row.get(other, 0.0) for other in existing_depts)
                else:
                    adj_score = 0.0
                rem_area = assignments[fl]['remaining_area']
                score = adj_score * 1000 + rem_area
                if (best_score is None) or (score > best_score):
                    best_score = score
                    best_floor = fl
            assigned_floor = best_floor

        # 3.2.d Place block if a floor was found
        if assigned_floor:
            fl = assigned_floor
            assignments[fl]['assigned_blocks'].append(blk)
            assignments[fl]['remaining_area'] -= blk_area
            assignments[fl]['remaining_capacity'] -= blk_capacity
            assignments[fl]['assigned_departments'].add(blk_dept)
            # Tally space‐mix area
            if blk_spacemix == 'ME':
                assignments[fl]['ME_area'] += blk_area
            elif blk_spacemix == 'WE':
                assignments[fl]['WE_area'] += blk_area
            elif blk_spacemix == 'US':
                assignments[fl]['US_area'] += blk_area
            elif blk_spacemix.lower() == 'support':
                assignments[fl]['Support_area'] += blk_area
            elif blk_spacemix.lower() == 'speciality':
                assignments[fl]['Speciality_area'] += blk_area
        else:
            print(f"Warning: Could not place typical block '{blk['Block_Name']}'")

    # 3.3 Build “Detailed” DataFrame directly from assigned row‐dicts
    assignment_list = []
    for fl, info in assignments.items():
        for blk in info['assigned_blocks']:
            assignment_list.append({
                'Floor': fl,
                'Department': blk['Department_Sub_Department'],
                'Block_Name': blk['Block_Name'],
                'Destination_Group': blk['Destination_Group'],
                'SpaceMix': blk['SpaceMix_(ME_WE_US_Support_Speciality)'],
                'Assigned_Area_SQM': blk['Cumulative_Area_SQM'],
                'Max_Occupancy': blk['Max_Occupancy_with_Capacity']
            })
    detailed_df = pd.DataFrame(assignment_list)

    # 3.4 Floor‐wise summary
    floor_summary_df = (
        detailed_df
        .groupby('Floor')
        .agg(
            Total_Blocks=('Block_Name', 'count'),
            Total_Area_SQM=('Assigned_Area_SQM', 'sum'),
            Total_Occupancy=('Max_Occupancy', 'sum')
        )
        .reset_index()
    )

    # 3.5 Space‐mix distribution per floor
    space_mix_df = (
        detailed_df
        .groupby(['Floor', 'SpaceMix'])
        .agg(Area_Per_SpaceMix=('Assigned_Area_SQM', 'sum'))
        .reset_index()
        .merge(floor_summary_df[['Floor', 'Total_Area_SQM']], on='Floor', how='left')
    )
    space_mix_df['SpaceMix_%'] = (
        space_mix_df['Area_Per_SpaceMix'] / space_mix_df['Total_Area_SQM'] * 100
    ).round(2)

    return detailed_df, floor_summary_df, space_mix_df

# ----------------------------------------
# Step 4: Generate & Export Excel Outputs
# ----------------------------------------

# 4.1 Centralized stacking
central_detailed, central_floor_sum, central_space_mix = run_stack_plan('centralized')

# 4.2 Decentralized stacking
decentral_detailed, decentral_floor_sum, decentral_space_mix = run_stack_plan('decentralized')

# 4.3 Write to Excel files
with pd.ExcelWriter('stack_plan_centralized.xlsx') as writer:
    central_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    central_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    central_space_mix.to_excel(writer, sheet_name='SpaceMix', index=False)

with pd.ExcelWriter('stack_plan_decentralized.xlsx') as writer:
    decentral_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    decentral_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    decentral_space_mix.to_excel(writer, sheet_name='SpaceMix', index=False)

print("✅ Generated Excel outputs:")
print("    • stack_plan_centralized.xlsx")
print("    • stack_plan_decentralized.xlsx")
```

```python
import pandas as pd
import random
import math

# ----------------------------------------
# Step 1: (Assuming all inputs are already loaded above)
# ----------------------------------------

# - all_floor_data
# - all_block_data
# - department_split_data
# - adjacency_data
# - De_Centralized_data

# ----------------------------------------
# Step 2: Preprocess block & department‐split
# ----------------------------------------

# 2.1 Convert cumulative area from SQFT to SQM
all_block_data['Cumulative_Area_SQM'] = all_block_data['Cumulative_Block_Circulation_Area_(SQFT)'] / 10.764

# 2.2 Split "Destination" vs. "Typical"
destination_blocks = all_block_data[all_block_data['Typical_Destination'] == 'Destination'].copy()
typical_blocks     = all_block_data[all_block_data['Typical_Destination'] == 'Typical'].copy()

# 2.3 Build dept_split_dict (though not used in this new logic for proportional assignment)
dept_split_dict = department_split_data.set_index('Department_Sub-Department')['Splittable'].to_dict()

# 2.4 Function to initialize floor assignments
def initialize_floor_assignments(floor_df):
    """
    Create a dict keyed by floor name. Each value tracks:
      - remaining_area       (uses column 'Usable_Area_(SQM)')
      - remaining_capacity   (uses column 'Max_Assignable_Floor_loading_Capacity')
      - assigned_blocks      (list of full block‐row dicts)
      - assigned_departments (set of sub‐departments)
      - ME_area, WE_area, US_area, Support_area, Speciality_area (for area tracking, not used for quota)
    """
    assignments = {}
    for _, row in floor_df.iterrows():
        floor = row['Name'].strip()
        assignments[floor] = {
            'remaining_area': row['Usable_Area_(SQM)'],
            'remaining_capacity': row['Max_Assignable_Floor_loading_Capacity'],
            'assigned_blocks': [],
            'assigned_departments': set(),
            'ME_area': 0.0,
            'WE_area': 0.0,
            'US_area': 0.0,
            'Support_area': 0.0,
            'Speciality_area': 0.0
        }
    return assignments

floors = list(initialize_floor_assignments(all_floor_data).keys())

# ----------------------------------------
# Step 3: Core stacking function with proportional “space‐mix” by unit counts
# ----------------------------------------

def run_stack_plan(mode):
    """
    mode: 'centralized' or 'decentralized'
    Returns three DataFrames:
      1) detailed_df      – block‐to‐floor assignments
      2) floor_summary_df – floor‐wise totals (block count, total area, total occupancy)
      3) space_mix_df     – floor‐wise % breakdown by unit counts per category
    """
    assignments = initialize_floor_assignments(all_floor_data)

    # 3.1 Helper: how many floors can each destination group occupy?
    def floors_allowed_for_dest(_group_name):
        if mode == 'centralized':
            return 2
        elif mode == 'decentralized':
            return len(floors)
        else:
            return len(floors)

    # 3.2 ASSIGN DESTINATION BLOCKS (unchanged from before)
    for group_name, group_df in destination_blocks.groupby('Destination_Group'):
        sorted_blocks = group_df.sort_values(by='Cumulative_Area_SQM', ascending=False).to_dict('records')
        max_floors = floors_allowed_for_dest(group_name)
        target_floors = floors[:max_floors]

        for blk in sorted_blocks:
            blk_area     = blk['Cumulative_Area_SQM']
            blk_capacity = blk['Max_Occupancy_with_Capacity']
            blk_dept     = blk['Department_Sub_Department'].strip()
            placed = False

            # 3.2.a Try allowed destination floors first
            for fl in target_floors:
                if assignments[fl]['remaining_area'] >= blk_area:
                    assignments[fl]['assigned_blocks'].append(blk)
                    assignments[fl]['remaining_area'] -= blk_area
                    assignments[fl]['remaining_capacity'] -= blk_capacity
                    assignments[fl]['assigned_departments'].add(blk_dept)
                    placed = True
                    break

            # 3.2.b If not placed, keep going up
            if not placed:
                for fl in floors[max_floors:]:
                    if assignments[fl]['remaining_area'] >= blk_area:
                        assignments[fl]['assigned_blocks'].append(blk)
                        assignments[fl]['remaining_area'] -= blk_area
                        assignments[fl]['remaining_capacity'] -= blk_capacity
                        assignments[fl]['assigned_departments'].add(blk_dept)
                        placed = True
                        break

            if not placed:
                print(f"Warning: Could not place destination block '{blk['Block_Name']}'")

    # 3.3 ASSIGN TYPICAL BLOCKS IN TWO PHASES:
    #      Phase A: ME blocks assigned randomly (subject to floor constraints).
    #      Phase B: For each other category (WE, US, Support, Speciality), assign proportionally
    #               in count according to the ME distribution per floor.

    # 3.3.A Phase A – Randomly assign ME blocks first
    me_blocks = [blk for blk in typical_blocks.to_dict('records') if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == 'ME']
    random.shuffle(me_blocks)

    for blk in me_blocks:
        blk_area     = blk['Cumulative_Area_SQM']
        blk_capacity = blk['Max_Occupancy_with_Capacity']
        blk_dept     = blk['Department_Sub_Department'].strip()

        # Build a list of floors in random order, then pick first that fits
        candidate_floors = floors.copy()
        random.shuffle(candidate_floors)
        placed = False
        for fl in candidate_floors:
            if assignments[fl]['remaining_area'] >= blk_area:
                assignments[fl]['assigned_blocks'].append(blk)
                assignments[fl]['remaining_area'] -= blk_area
                assignments[fl]['remaining_capacity'] -= blk_capacity
                assignments[fl]['assigned_departments'].add(blk_dept)
                assignments[fl]['ME_area'] += blk_area
                placed = True
                break
        if not placed:
            print(f"Warning: Could not place ME block '{blk['Block_Name']}'")

    # 3.3.B Count how many ME blocks ended up on each floor
    me_count_per_floor = {fl: 0 for fl in floors}
    for fl, info in assignments.items():
        me_count_per_floor[fl] = sum(
            1 for blk in info['assigned_blocks']
            if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == 'ME'
        )
    total_me = sum(me_count_per_floor.values())
    # If no ME blocks were placed (total_me == 0), avoid division by zero; distribute equally
    if total_me == 0:
        me_frac_per_floor = {fl: 1/len(floors) for fl in floors}
    else:
        me_frac_per_floor = {fl: me_count_per_floor[fl] / total_me for fl in floors}

    # 3.3.C For each other category, assign proportionally by count
    other_categories = ['WE', 'US', 'Support', 'Speciality']
    for category in other_categories:
        cat_blocks = [
            blk for blk in typical_blocks.to_dict('records')
            if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == category
        ]
        total_cat = len(cat_blocks)
        if total_cat == 0:
            continue

        # Compute integer target per floor: round(frac * total_cat) initially
        raw_targets = {fl: me_frac_per_floor[fl] * total_cat for fl in floors}
        target_counts = {fl: int(round(raw_targets[fl])) for fl in floors}

        # Adjust rounding so sum of target_counts equals total_cat
        diff = total_cat - sum(target_counts.values())
        if diff != 0:
            # Compute floors sorted by descending fractional part to add/subtract
            fractional_parts = {
                fl: raw_targets[fl] - math.floor(raw_targets[fl]) for fl in floors
            }
            # If we need to add some counts
            if diff > 0:
                # Assign +1 to floors with largest fractional part first
                for fl in sorted(floors, key=lambda x: fractional_parts[x], reverse=True)[:diff]:
                    target_counts[fl] += 1
            else:
                # diff < 0: need to remove abs(diff) counts; remove from floors with smallest fractional part
                for fl in sorted(floors, key=lambda x: fractional_parts[x])[: -diff]:
                    target_counts[fl] -= 1

        # Now assign blocks category‐wise
        # Shuffle cat_blocks to avoid ordering bias
        random.shuffle(cat_blocks)
        # Track how many assigned so far per floor for this category
        assigned_counts = {fl: 0 for fl in floors}

        # For each block, attempt to place on any floor where assigned_counts[f] < target_counts[f]
        for blk in cat_blocks:
            blk_area     = blk['Cumulative_Area_SQM']
            blk_capacity = blk['Max_Occupancy_with_Capacity']
            blk_dept     = blk['Department_Sub_Department'].strip()

            placed = False
            # Build list of floors sorted by how far they are from target (descending)
            deficits = {fl: target_counts[fl] - assigned_counts[fl] for fl in floors}
            # Only consider floors with positive deficit first
            floors_with_deficit = [fl for fl, d in deficits.items() if d > 0]
            # If no deficits remain, we'll fallback to any floor that fits
            if floors_with_deficit:
                candidate_floors = sorted(
                    floors_with_deficit,
                    key=lambda x: deficits[x],
                    reverse=True
                )
            else:
                candidate_floors = floors.copy()

            for fl in candidate_floors:
                if assignments[fl]['remaining_area'] >= blk_area:
                    assignments[fl]['assigned_blocks'].append(blk)
                    assignments[fl]['remaining_area'] -= blk_area
                    assignments[fl]['remaining_capacity'] -= blk_capacity
                    assignments[fl]['assigned_departments'].add(blk_dept)
                    # Track space‐mix area for reporting (not used for quotas)
                    if category == 'WE':
                        assignments[fl]['WE_area'] += blk_area
                    elif category == 'US':
                        assignments[fl]['US_area'] += blk_area
                    elif category == 'Support':
                        assignments[fl]['Support_area'] += blk_area
                    elif category == 'Speciality':
                        assignments[fl]['Speciality_area'] += blk_area

                    # Increment assigned count for this category on that floor
                    assigned_counts[fl] += 1
                    placed = True
                    break

            if not placed:
                # If no floor with positive deficit can fit it, try ANY floor in random order
                fallback = floors.copy()
                random.shuffle(fallback)
                for fl in fallback:
                    if assignments[fl]['remaining_area'] >= blk_area:
                        assignments[fl]['assigned_blocks'].append(blk)
                        assignments[fl]['remaining_area'] -= blk_area
                        assignments[fl]['remaining_capacity'] -= blk_capacity
                        assignments[fl]['assigned_departments'].add(blk_dept)
                        if category == 'WE':
                            assignments[fl]['WE_area'] += blk_area
                        elif category == 'US':
                            assignments[fl]['US_area'] += blk_area
                        elif category == 'Support':
                            assignments[fl]['Support_area'] += blk_area
                        elif category == 'Speciality':
                            assignments[fl]['Speciality_area'] += blk_area
                        assigned_counts[fl] += 1
                        placed = True
                        break

            if not placed:
                print(f"Warning: Could not place {category} block '{blk['Block_Name']}'")

    # 3.4 BUILD “Detailed” DataFrame
    assignment_list = []
    for fl, info in assignments.items():
        for blk in info['assigned_blocks']:
            assignment_list.append({
                'Floor': fl,
                'Department': blk['Department_Sub_Department'],
                'Block_Name': blk['Block_Name'],
                'Destination_Group': blk['Destination_Group'],
                'SpaceMix': blk['SpaceMix_(ME_WE_US_Support_Speciality)'],
                'Assigned_Area_SQM': blk['Cumulative_Area_SQM'],
                'Max_Occupancy': blk['Max_Occupancy_with_Capacity']
            })
    detailed_df = pd.DataFrame(assignment_list)

    # 3.5 Floor‐wise summary
    floor_summary_df = (
        detailed_df
        .groupby('Floor')
        .agg(
            Total_Blocks=('Block_Name', 'count'),
            Total_Area_SQM=('Assigned_Area_SQM', 'sum'),
            Total_Occupancy=('Max_Occupancy', 'sum')
        )
        .reset_index()
    )

    # 3.6 Space‐mix breakdown by UNIT COUNT per floor
    # Count how many blocks of each category landed on each floor
    rows = []
    for fl, info in assignments.items():
        counts = {'ME': 0, 'WE': 0, 'US': 0, 'Support': 0, 'Speciality': 0}
        for blk in info['assigned_blocks']:
            cat = blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip()
            if cat in counts:
                counts[cat] += 1
        total_blocks_floor = sum(counts.values())
        if total_blocks_floor == 0:
            continue
        for cat, cnt in counts.items():
            pct = cnt / total_blocks_floor * 100
            rows.append({
                'Floor': fl,
                'SpaceMix': cat,
                'Unit_Count': cnt,
                'Percent_of_Floor': round(pct, 2)
            })

    space_mix_df = pd.DataFrame(rows)

    return detailed_df, floor_summary_df, space_mix_df

# ----------------------------------------
# Step 4: Generate & Export Excel Outputs
# ----------------------------------------

# 4.1 Centralized stacking
central_detailed, central_floor_sum, central_space_mix = run_stack_plan('centralized')

# 4.2 Decentralized stacking
decentral_detailed, decentral_floor_sum, decentral_space_mix = run_stack_plan('decentralized')

# 4.3 Write to Excel files
with pd.ExcelWriter('stack_plan_centralized1.xlsx') as writer:
    central_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    central_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    central_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

with pd.ExcelWriter('stack_plan_decentralized1.xlsx') as writer:
    decentral_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    decentral_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    decentral_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

print("✅ Generated Excel outputs with proportional unit‐based space‐mix:")
print("    • stack_plan_centralized1.xlsx")
print("    • stack_plan_decentralized1.xlsx")
```

```python
import pandas as pd
import random
import math

# ----------------------------------------
# Step 1: Assume the following are already loaded:
#   - all_floor_data        (floors sheet)
#   - all_block_data        (blocks sheet)
#   - department_split_data (department split sheet)
#   - adjacency_data        (adjacency matrix)
#   - De_Centralized_data   (decentralized logic)
#
# Column names must match exactly:
#   Floors sheet:      'Name', 'Usable_Area_(SQM)', 'Max_Assignable_Floor_loading_Capacity'
#   Blocks sheet:      'Block_ID', 'Block_Name', 'Block_Instance_Name', 'Destination_Group',
#                      'Block_Type', 'Department_Sub_Department', 'Block_Area (SQFT)',
#                      'Block_Capacity', 'Typical_Destination', 'Cumulative_Block_Circulation_Area_(SQFT)',
#                      'Max_Occupancy_with_Capacity', 'SpaceMix_(ME_WE_US_Support_Speciality)'
#   Department Split:  'BU_Department_Sub-Department', 'Splittable'
#   Adjacency sheet:   square matrix indexed by sub‐department names
#   De-Centralized Logic: rows indicating how many floors each destination group may occupy
# ----------------------------------------

# Precompute "Cumulative_Area_SQM" for each block
all_block_data['Cumulative_Area_SQM'] = all_block_data['Cumulative_Block_Circulation_Area_(SQFT)'] / 10.764

# Split out Destination vs. Typical
destination_blocks = all_block_data[all_block_data['Typical_Destination'] == 'Destination'].copy()
typical_blocks     = all_block_data[all_block_data['Typical_Destination'] == 'Typical'].copy()

# Dept-split dict (not used here, but we keep for completeness)
department_split_data = department_split_data.rename(
    columns={'BU_Department_Sub-Department': 'Department_Sub-Department'}
)
dept_split_dict = department_split_data.set_index('Department_Sub-Department')['Splittable'].to_dict()

# Helper to initialize floor assignments
def initialize_floor_assignments(floor_df):
    """
    Returns a dict keyed by floor name. Each entry contains:
      - remaining_area
      - remaining_capacity
      - assigned_blocks (list of block‐row dicts)
      - assigned_departments (set of sub‐departments)
      - ME_area, WE_area, US_area, Support_area, Speciality_area (floats)
    """
    assignments = {}
    for _, row in floor_df.iterrows():
        floor = row['Name'].strip()
        assignments[floor] = {
            'remaining_area': row['Usable_Area_(SQM)'],
            'remaining_capacity': row['Max_Assignable_Floor_loading_Capacity'],
            'assigned_blocks': [],
            'assigned_departments': set(),
            'ME_area': 0.0,
            'WE_area': 0.0,
            'US_area': 0.0,
            'Support_area': 0.0,
            'Speciality_area': 0.0
        }
    return assignments

floors = list(initialize_floor_assignments(all_floor_data).keys())

# ----------------------------------------
# Step 2: Core stacking function with proportional unit‐based space‐mix
# ----------------------------------------

def run_stack_plan(mode):
    """
    Build a stack plan under 'centralized' or 'decentralized' rules, assigning ME blocks first
    then other categories in proportion to the ME distribution per floor.
    Returns three DataFrames:
      1) detailed_df      – each block’s assigned floor
      2) floor_summary_df – floor‐wise totals (block count, total area, total occupancy)
      3) space_mix_df     – floor‐wise % breakdown by unit count per category
    """
    assignments = initialize_floor_assignments(all_floor_data)

    # Helper: how many floors each destination group can occupy?
    def floors_allowed_for_dest(_group_name):
        if mode == 'centralized':
            return 2
        elif mode == 'decentralized':
            return len(floors)
        else:
            return len(floors)

    # ----- Phase 1: Assign DESTINATION blocks exactly as before -----
    for group_name, group_df in destination_blocks.groupby('Destination_Group'):
        sorted_blocks = group_df.sort_values(by='Cumulative_Area_SQM', ascending=False).to_dict('records')
        max_floors = floors_allowed_for_dest(group_name)
        target_floors = floors[:max_floors]

        for blk in sorted_blocks:
            blk_area     = blk['Cumulative_Area_SQM']
            blk_capacity = blk['Max_Occupancy_with_Capacity']
            blk_dept     = blk['Department_Sub_Department'].strip()
            placed = False

            # Try only the allowed destination floors
            for fl in target_floors:
                if assignments[fl]['remaining_area'] >= blk_area:
                    assignments[fl]['assigned_blocks'].append(blk)
                    assignments[fl]['remaining_area'] -= blk_area
                    assignments[fl]['remaining_capacity'] -= blk_capacity
                    assignments[fl]['assigned_departments'].add(blk_dept)
                    placed = True
                    break

            if not placed:
                # If no destination floor fits, move up
                for fl in floors[max_floors:]:
                    if assignments[fl]['remaining_area'] >= blk_area:
                        assignments[fl]['assigned_blocks'].append(blk)
                        assignments[fl]['remaining_area'] -= blk_area
                        assignments[fl]['remaining_capacity'] -= blk_capacity
                        assignments[fl]['assigned_departments'].add(blk_dept)
                        placed = True
                        break

            if not placed:
                print(f"Warning: Could not place destination block '{blk['Block_Name']}'")

    # ----- Phase 2A: Assign ME blocks at random (subject to floor constraints) -----
    me_blocks = [
        blk for blk in typical_blocks.to_dict('records')
        if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == 'ME'
    ]
    random.shuffle(me_blocks)

    for blk in me_blocks:
        blk_area     = blk['Cumulative_Area_SQM']
        blk_capacity = blk['Max_Occupancy_with_Capacity']
        blk_dept     = blk['Department_Sub_Department'].strip()

        candidate_floors = floors.copy()
        random.shuffle(candidate_floors)
        placed = False
        for fl in candidate_floors:
            if assignments[fl]['remaining_area'] >= blk_area:
                assignments[fl]['assigned_blocks'].append(blk)
                assignments[fl]['remaining_area'] -= blk_area
                assignments[fl]['remaining_capacity'] -= blk_capacity
                assignments[fl]['assigned_departments'].add(blk_dept)
                assignments[fl]['ME_area'] += blk_area
                placed = True
                break
        if not placed:
            print(f"Warning: Could not place ME block '{blk['Block_Name']}'")

    # ----- Compute ME distribution per floor (unit counts) -----
    me_count_per_floor = {fl: 0 for fl in floors}
    for fl, info in assignments.items():
        me_count_per_floor[fl] = sum(
            1 for blk in info['assigned_blocks']
            if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == 'ME'
        )
    total_me = sum(me_count_per_floor.values())
    if total_me == 0:
        # If no ME blocks placed, distribute fractions equally
        me_frac_per_floor = {fl: 1 / len(floors) for fl in floors}
    else:
        me_frac_per_floor = {
            fl: me_count_per_floor[fl] / total_me
            for fl in floors
        }

    # ----- Phase 2B: Assign OTHER categories proportionally to ME fractions -----
    other_categories = ['WE', 'US', 'Support', 'Speciality']

    for category in other_categories:
        cat_blocks = [
            blk for blk in typical_blocks.to_dict('records')
            if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == category
        ]
        total_cat = len(cat_blocks)
        if total_cat == 0:
            continue

        # Compute raw targets per floor = me_frac_per_floor * total_cat
        raw_targets = {fl: me_frac_per_floor[fl] * total_cat for fl in floors}
        target_counts = {fl: int(round(raw_targets[fl])) for fl in floors}

        # Adjust rounding so sum == total_cat
        diff = total_cat - sum(target_counts.values())
        if diff != 0:
            fractional_parts = {fl: raw_targets[fl] - math.floor(raw_targets[fl]) for fl in floors}
            if diff > 0:
                # Add one to floors with largest fractional parts
                for fl in sorted(floors, key=lambda x: fractional_parts[x], reverse=True)[:diff]:
                    target_counts[fl] += 1
            else:
                # Remove one from floors with smallest fractional parts
                for fl in sorted(floors, key=lambda x: fractional_parts[x])[: -diff]:
                    target_counts[fl] -= 1

        # Shuffle category blocks to avoid ordering bias
        random.shuffle(cat_blocks)
        assigned_counts = {fl: 0 for fl in floors}

        for blk in cat_blocks:
            blk_area     = blk['Cumulative_Area_SQM']
            blk_capacity = blk['Max_Occupancy_with_Capacity']
            blk_dept     = blk['Department_Sub_Department'].strip()

            # Build list of floors with positive deficit first
            deficits = {fl: target_counts[fl] - assigned_counts[fl] for fl in floors}
            floors_with_deficit = [fl for fl, d in deficits.items() if d > 0]

            if floors_with_deficit:
                candidate_floors = sorted(
                    floors_with_deficit,
                    key=lambda x: deficits[x],
                    reverse=True
                )
            else:
                candidate_floors = floors.copy()

            placed = False
            for fl in candidate_floors:
                if assignments[fl]['remaining_area'] >= blk_area:
                    assignments[fl]['assigned_blocks'].append(blk)
                    assignments[fl]['remaining_area'] -= blk_area
                    assignments[fl]['remaining_capacity'] -= blk_capacity
                    assignments[fl]['assigned_departments'].add(blk_dept)
                    if category == 'WE':
                        assignments[fl]['WE_area'] += blk_area
                    elif category == 'US':
                        assignments[fl]['US_area'] += blk_area
                    elif category == 'Support':
                        assignments[fl]['Support_area'] += blk_area
                    elif category == 'Speciality':
                        assignments[fl]['Speciality_area'] += blk_area
                    assigned_counts[fl] += 1
                    placed = True
                    break

            if not placed:
                # Fallback: try any floor in random order
                fallback_floors = floors.copy()
                random.shuffle(fallback_floors)
                for fl in fallback_floors:
                    if assignments[fl]['remaining_area'] >= blk_area:
                        assignments[fl]['assigned_blocks'].append(blk)
                        assignments[fl]['remaining_area'] -= blk_area
                        assignments[fl]['remaining_capacity'] -= blk_capacity
                        assignments[fl]['assigned_departments'].add(blk_dept)
                        if category == 'WE':
                            assignments[fl]['WE_area'] += blk_area
                        elif category == 'US':
                            assignments[fl]['US_area'] += blk_area
                        elif category == 'Support':
                            assignments[fl]['Support_area'] += blk_area
                        elif category == 'Speciality':
                            assignments[fl]['Speciality_area'] += blk_area
                        assigned_counts[fl] += 1
                        placed = True
                        break

            if not placed:
                print(f"Warning: Could not place {category} block '{blk['Block_Name']}'")

    # ----- Phase 3: Build Detailed & Summary DataFrames -----
    assignment_list = []
    for fl, info in assignments.items():
        for blk in info['assigned_blocks']:
            assignment_list.append({
                'Floor': fl,
                'Department': blk['Department_Sub_Department'],
                'Block_Name': blk['Block_Name'],
                'Destination_Group': blk['Destination_Group'],
                'SpaceMix': blk['SpaceMix_(ME_WE_US_Support_Speciality)'],
                'Assigned_Area_SQM': blk['Cumulative_Area_SQM'],
                'Max_Occupancy': blk['Max_Occupancy_with_Capacity']
            })
    detailed_df = pd.DataFrame(assignment_list)

    # Floor‐wise summary
    floor_summary_df = (
        detailed_df
        .groupby('Floor')
        .agg(
            Total_Blocks=('Block_Name', 'count'),
            Total_Area_SQM=('Assigned_Area_SQM', 'sum'),
            Total_Occupancy=('Max_Occupancy', 'sum')
        )
        .reset_index()
    )

    # Space‐mix percentages by unit count per floor
    rows = []
    for fl, info in assignments.items():
        counts = {'ME': 0, 'WE': 0, 'US': 0, 'Support': 0, 'Speciality': 0}
        for blk in info['assigned_blocks']:
            cat = blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip()
            if cat in counts:
                counts[cat] += 1
        total_blocks_floor = sum(counts.values())
        if total_blocks_floor == 0:
            continue
        for cat, cnt in counts.items():
            pct = cnt / total_blocks_floor * 100
            # But the user wants: pct = cnt / total_cat_overall × 100
            # We'll also compute that in a separate column.
            total_cat_overall = len(typical_blocks[typical_blocks['SpaceMix_(ME_WE_US_Support_Speciality)'].str.strip() == cat])
            if total_cat_overall == 0:
                pct_overall = 0.0
            else:
                pct_overall = cnt / total_cat_overall * 100
            rows.append({
                'Floor': fl,
                'SpaceMix': cat,
                'Unit_Count_on_Floor': cnt,
                'Pct_of_Floor_UC': round(pct, 2),
                'Pct_of_Overall_UC': round(pct_overall, 2)
            })

    space_mix_df = pd.DataFrame(rows)

    return detailed_df, floor_summary_df, space_mix_df

# ----------------------------------------
# Step 4: Generate & Export Excel Outputs
# ----------------------------------------

central_detailed, central_floor_sum, central_space_mix = run_stack_plan('centralized')
decentral_detailed, decentral_floor_sum, decentral_space_mix = run_stack_plan('decentralized')

with pd.ExcelWriter('stack_plan_centralized2.xlsx') as writer:
    central_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    central_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    central_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

with pd.ExcelWriter('stack_plan_decentralized2.xlsx') as writer:
    decentral_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    decentral_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    decentral_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

print("✅ Generated Excel outputs with unit‐based space‐mix percentages:")
print("    • stack_plan_centralized2.xlsx")
print("    • stack_plan_decentralized2.xlsx")
```

```python
import pandas as pd
import random
import math

# ----------------------------------------
# Step 1: Load Input Sheets
# ----------------------------------------

excel_path = '/content/AR-1.xlsx'  # ← adjust this path if needed

# 1.1 Floors sheet
all_floor_data = pd.read_excel(excel_path, sheet_name='Program Table Input 2 - Floor')
all_floor_data.columns = all_floor_data.columns.str.strip()

# 1.2 Blocks sheet
all_block_data = pd.read_excel(excel_path, sheet_name='Program Table Input 1 - Block')
all_block_data.columns = all_block_data.columns.str.strip()

# 1.3 Department Split sheet (not used for final space‐mix logic, but loaded for completeness)
department_split_data = pd.read_excel(excel_path, sheet_name='Department Split', skiprows=1)
department_split_data.columns = department_split_data.columns.str.strip()
department_split_data = department_split_data.rename(
    columns={'BU_Department_Sub-Department': 'Department_Sub-Department'}
)

# 1.4 Adjacency sheet
xls = pd.ExcelFile(excel_path)
adjacency_sheet_name = [name for name in xls.sheet_names if "Adjacency" in name][0]
raw_data = xls.parse(adjacency_sheet_name, header=1, index_col=0)
adjacency_data = raw_data.apply(pd.to_numeric, errors='coerce')
adjacency_data.index   = adjacency_data.index.str.strip()
adjacency_data.columns = adjacency_data.columns.str.strip()

# 1.5 De-Centralized Logic sheet
df_logic = pd.read_excel(excel_path, sheet_name='De-Centralized Logic', header=None)
De_Centralized_data = {}
current_section = None
for _, row in df_logic.iterrows():
    first_cell = str(row[0]).strip() if pd.notna(row[0]) else ""
    if first_cell in ["Centralised", "Semi Centralized", "DeCentralised"]:
        current_section = first_cell
        De_Centralized_data[current_section] = {}
    elif current_section and first_cell == "Destination":
        De_Centralized_data[current_section]["Destination"] = int(row[1]) if pd.notna(row[1]) else None

# ----------------------------------------
# Step 2: Preprocess Blocks & Department Split
# ----------------------------------------

# 2.1 Convert cumulative circulation area from SQFT → SQM
all_block_data['Cumulative_Area_SQM'] = (
    all_block_data['Cumulative_Block_Circulation_Area_(SQFT)'] / 10.764
)

# 2.2 Separate Destination vs. Typical blocks
destination_blocks = all_block_data[all_block_data['Typical_Destination'] == 'Destination'].copy()
typical_blocks     = all_block_data[all_block_data['Typical_Destination'] == 'Typical'].copy()

# 2.3 Dept‐split dict (not needed for space‐mix but loaded)
dept_split_dict = department_split_data.set_index('Department_Sub-Department')['Splittable'].to_dict()

# ----------------------------------------
# Step 3: Initialize Floor Assignments
# ----------------------------------------

def initialize_floor_assignments(floor_df):
    """
    Returns a dict keyed by floor name. Each entry contains:
      - remaining_area
      - remaining_capacity
      - assigned_blocks      (list of block‐row dicts)
      - assigned_departments (set of sub‐departments)
      - ME_area, WE_area, US_area, Support_area, Speciality_area (floats, for area tracking)
    """
    assignments = {}
    for _, row in floor_df.iterrows():
        floor = row['Name'].strip()
        assignments[floor] = {
            'remaining_area': row['Usable_Area_(SQM)'],
            'remaining_capacity': row['Max_Assignable_Floor_loading_Capacity'],
            'assigned_blocks': [],
            'assigned_departments': set(),
            'ME_area': 0.0,
            'WE_area': 0.0,
            'US_area': 0.0,
            'Support_area': 0.0,
            'Speciality_area': 0.0
        }
    return assignments

floors = list(initialize_floor_assignments(all_floor_data).keys())

# ----------------------------------------
# Step 4: Core Stacking Function
# ----------------------------------------

def run_stack_plan(mode):
    """
    mode: 'centralized' or 'decentralized'
    Returns:
      1) detailed_df      – each block’s assigned floor, department, block name, destination group, space mix, area (SQM), occupancy
      2) floor_summary_df – for each floor: total number of blocks, total area assigned, total occupancy
      3) space_mix_df     – for each floor and each category among {ME, WE, US, Support, Speciality}:
                             - Unit_Count_on_Floor
                             - Pct_of_Floor_UC      = (blocks_of_category_on_floor / total_blocks_on_that_floor) × 100%
                             - Pct_of_Overall_UC    = (blocks_of_category_on_floor / total_blocks_of_category_overall) × 100%
    """
    assignments = initialize_floor_assignments(all_floor_data)

    def floors_allowed_for_dest(_group_name):
        if mode == 'centralized':
            return 2
        elif mode == 'decentralized':
            return len(floors)
        else:
            return len(floors)

    # Phase 1: Assign Destination Blocks (largest‐area first within each Destination_Group)
    for group_name, group_df in destination_blocks.groupby('Destination_Group'):
        sorted_blocks = group_df.sort_values(by='Cumulative_Area_SQM', ascending=False).to_dict('records')
        max_floors = floors_allowed_for_dest(group_name)
        target_floors = floors[:max_floors]

        for blk in sorted_blocks:
            blk_area     = blk['Cumulative_Area_SQM']
            blk_capacity = blk['Max_Occupancy_with_Capacity']
            blk_dept     = blk['Department_Sub_Department'].strip()
            placed = False

            # First try only the “destination” floors
            for fl in target_floors:
                if assignments[fl]['remaining_area'] >= blk_area:
                    assignments[fl]['assigned_blocks'].append(blk)
                    assignments[fl]['remaining_area'] -= blk_area
                    assignments[fl]['remaining_capacity'] -= blk_capacity
                    assignments[fl]['assigned_departments'].add(blk_dept)
                    placed = True
                    break

            # If none of those fit, keep going up
            if not placed:
                for fl in floors[max_floors:]:
                    if assignments[fl]['remaining_area'] >= blk_area:
                        assignments[fl]['assigned_blocks'].append(blk)
                        assignments[fl]['remaining_area'] -= blk_area
                        assignments[fl]['remaining_capacity'] -= blk_capacity
                        assignments[fl]['assigned_departments'].add(blk_dept)
                        placed = True
                        break

            if not placed:
                print(f"Warning: Could not place destination block '{blk['Block_Name']}'")

    # Phase 2A: Randomly assign all ME blocks (typical)
    me_blocks = [
        blk for blk in typical_blocks.to_dict('records')
        if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == 'ME'
    ]
    random.shuffle(me_blocks)
    for blk in me_blocks:
        blk_area     = blk['Cumulative_Area_SQM']
        blk_capacity = blk['Max_Occupancy_with_Capacity']
        blk_dept     = blk['Department_Sub_Department'].strip()

        candidate_floors = floors.copy()
        random.shuffle(candidate_floors)
        placed = False
        for fl in candidate_floors:
            if assignments[fl]['remaining_area'] >= blk_area:
                assignments[fl]['assigned_blocks'].append(blk)
                assignments[fl]['remaining_area'] -= blk_area
                assignments[fl]['remaining_capacity'] -= blk_capacity
                assignments[fl]['assigned_departments'].add(blk_dept)
                assignments[fl]['ME_area'] += blk_area
                placed = True
                break
        if not placed:
            print(f"Warning: Could not place ME block '{blk['Block_Name']}'")

    # Compute ME distribution per floor
    me_count_per_floor = {fl: 0 for fl in floors}
    for fl, info in assignments.items():
        me_count_per_floor[fl] = sum(
            1 for blk in info['assigned_blocks']
            if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == 'ME'
        )
    total_me = sum(me_count_per_floor.values())
    if total_me == 0:
        # If no ME blocks placed, treat all floors equally
        me_frac_per_floor = {fl: 1 / len(floors) for fl in floors}
    else:
        me_frac_per_floor = {
            fl: me_count_per_floor[fl] / total_me
            for fl in floors
        }

    # Phase 2B: Assign all other categories proportionally
    other_categories = ['WE', 'US', 'Support', 'Speciality']
    for category in other_categories:
        cat_blocks = [
            blk for blk in typical_blocks.to_dict('records')
            if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == category
        ]
        total_cat = len(cat_blocks)
        if total_cat == 0:
            continue

        # Compute raw target count = me_frac_per_floor * total_cat
        raw_targets = {fl: me_frac_per_floor[fl] * total_cat for fl in floors}
        target_counts = {fl: int(round(raw_targets[fl])) for fl in floors}

        # Adjust rounding error so sum(target_counts) == total_cat
        diff = total_cat - sum(target_counts.values())
        if diff != 0:
            fractional_parts = {
                fl: raw_targets[fl] - math.floor(raw_targets[fl]) for fl in floors
            }
            if diff > 0:
                # Increase floors with largest fractional part
                for fl in sorted(floors, key=lambda x: fractional_parts[x], reverse=True)[:diff]:
                    target_counts[fl] += 1
            else:
                # Decrease floors with smallest fractional part
                for fl in sorted(floors, key=lambda x: fractional_parts[x])[: -diff]:
                    target_counts[fl] -= 1

        random.shuffle(cat_blocks)
        assigned_counts = {fl: 0 for fl in floors}

        for blk in cat_blocks:
            blk_area     = blk['Cumulative_Area_SQM']
            blk_capacity = blk['Max_Occupancy_with_Capacity']
            blk_dept     = blk['Department_Sub_Department'].strip()

            # Choose floors that still need category‐units first
            deficits = {fl: target_counts[fl] - assigned_counts[fl] for fl in floors}
            floors_with_deficit = [fl for fl, d in deficits.items() if d > 0]
            if floors_with_deficit:
                candidate_floors = sorted(
                    floors_with_deficit,
                    key=lambda x: deficits[x],
                    reverse=True
                )
            else:
                candidate_floors = floors.copy()

            placed = False
            for fl in candidate_floors:
                if assignments[fl]['remaining_area'] >= blk_area:
                    assignments[fl]['assigned_blocks'].append(blk)
                    assignments[fl]['remaining_area'] -= blk_area
                    assignments[fl]['remaining_capacity'] -= blk_capacity
                    assignments[fl]['assigned_departments'].add(blk_dept)
                    if category == 'WE':
                        assignments[fl]['WE_area'] += blk_area
                    elif category == 'US':
                        assignments[fl]['US_area'] += blk_area
                    elif category == 'Support':
                        assignments[fl]['Support_area'] += blk_area
                    elif category == 'Speciality':
                        assignments[fl]['Speciality_area'] += blk_area
                    assigned_counts[fl] += 1
                    placed = True
                    break

            if not placed:
                # Fallback: try any floor in random order
                fallback = floors.copy()
                random.shuffle(fallback)
                for fl in fallback:
                    if assignments[fl]['remaining_area'] >= blk_area:
                        assignments[fl]['assigned_blocks'].append(blk)
                        assignments[fl]['remaining_area'] -= blk_area
                        assignments[fl]['remaining_capacity'] -= blk_capacity
                        assignments[fl]['assigned_departments'].add(blk_dept)
                        if category == 'WE':
                            assignments[fl]['WE_area'] += blk_area
                        elif category == 'US':
                            assignments[fl]['US_area'] += blk_area
                        elif category == 'Support':
                            assignments[fl]['Support_area'] += blk_area
                        elif category == 'Speciality':
                            assignments[fl]['Speciality_area'] += blk_area
                        assigned_counts[fl] += 1
                        placed = True
                        break

            if not placed:
                print(f"Warning: Could not place {category} block '{blk['Block_Name']}'")

    # Phase 3: Build Detailed & Summary DataFrames

    # 3.1 “Detailed” DataFrame
    assignment_list = []
    for fl, info in assignments.items():
        for blk in info['assigned_blocks']:
            assignment_list.append({
                'Floor': fl,
                'Department': blk['Department_Sub_Department'],
                'Block_Name': blk['Block_Name'],
                'Destination_Group': blk['Destination_Group'],
                'SpaceMix': blk['SpaceMix_(ME_WE_US_Support_Speciality)'],
                'Assigned_Area_SQM': blk['Cumulative_Area_SQM'],
                'Max_Occupancy': blk['Max_Occupancy_with_Capacity']
            })
    detailed_df = pd.DataFrame(assignment_list)

    # 3.2 “Floor_Summary” DataFrame
    floor_summary_df = (
        detailed_df
        .groupby('Floor')
        .agg(
            Total_Blocks=('Block_Name', 'count'),
            Total_Area_SQM=('Assigned_Area_SQM', 'sum'),
            Total_Occupancy=('Max_Occupancy', 'sum')
        )
        .reset_index()
    )

    # 3.3 “SpaceMix_By_Units” DataFrame
    all_categories = ['ME', 'WE', 'US', 'Support', 'Speciality']
    category_totals = {
        cat: len(typical_blocks[
            typical_blocks['SpaceMix_(ME_WE_US_Support_Speciality)'].str.strip() == cat
        ])
        for cat in all_categories
    }

    rows = []
    for fl, info in assignments.items():
        counts = {cat: 0 for cat in all_categories}
        for blk in info['assigned_blocks']:
            cat = blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip()
            if cat in counts:
                counts[cat] += 1
        total_blocks_on_floor = sum(counts.values())

        for cat in all_categories:
            cnt = counts[cat]
            # Percent of floor’s blocks
            pct_of_floor = (cnt / total_blocks_on_floor * 100) if total_blocks_on_floor else 0.0
            # Percent of overall blocks of that category
            total_cat = category_totals[cat]
            pct_overall = (cnt / total_cat * 100) if total_cat else 0.0

            rows.append({
                'Floor': fl,
                'SpaceMix': cat,
                'Unit_Count_on_Floor': cnt,
                'Pct_of_Floor_UC': round(pct_of_floor, 2),
                'Pct_of_Overall_UC': round(pct_overall, 2)
            })

    space_mix_df = pd.DataFrame(rows)

    return detailed_df, floor_summary_df, space_mix_df

# ----------------------------------------
# Step 5: Generate & Export Excel Outputs
# ----------------------------------------

central_detailed, central_floor_sum, central_space_mix = run_stack_plan('centralized')
decentral_detailed, decentral_floor_sum, decentral_space_mix = run_stack_plan('decentralized')

with pd.ExcelWriter('stack_plan_centralized4.xlsx') as writer:
    central_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    central_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    central_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

with pd.ExcelWriter('stack_plan_decentralized4.xlsx') as writer:
    decentral_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    decentral_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    decentral_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

print("✅ Generated Excel outputs with US & Speciality included in space‐mix:")
print("    • stack_plan_centralized4.xlsx")
print("    • stack_plan_decentralized4.xlsx")
```

```python
import pandas as pd
import random
import math

# ----------------------------------------
# Step 1: Load Input Sheets
# ----------------------------------------

excel_path = '/content/AR-1.xlsx'  # ← adjust if needed

# 1.1 Floors sheet
all_floor_data = pd.read_excel(excel_path, sheet_name='Program Table Input 2 - Floor')
all_floor_data.columns = all_floor_data.columns.str.strip()

# 1.2 Blocks sheet
all_block_data = pd.read_excel(excel_path, sheet_name='Program Table Input 1 - Block')
all_block_data.columns = all_block_data.columns.str.strip()

# 1.3 Department Split sheet
department_split_data = pd.read_excel(excel_path, sheet_name='Department Split', skiprows=1)
department_split_data.columns = department_split_data.columns.str.strip()
department_split_data = department_split_data.rename(
    columns={'BU_Department_Sub-Department': 'Department_Sub-Department'}
)

# 1.4 Adjacency sheet
xls = pd.ExcelFile(excel_path)
adjacency_sheet_name = [name for name in xls.sheet_names if "Adjacency" in name][0]
raw_data = xls.parse(adjacency_sheet_name, header=1, index_col=0)
adjacency_data = raw_data.apply(pd.to_numeric, errors='coerce')
adjacency_data.index = adjacency_data.index.str.strip()
adjacency_data.columns = adjacency_data.columns.str.strip()

# 1.5 De-Centralized Logic sheet
df_logic = pd.read_excel(excel_path, sheet_name='De-Centralized Logic', header=None)
De_Centralized_data = {}
current_section = None
for _, row in df_logic.iterrows():
    first_cell = str(row[0]).strip() if pd.notna(row[0]) else ""
    if first_cell in ["Centralised", "Semi Centralized", "DeCentralised"]:
        current_section = first_cell
        De_Centralized_data[current_section] = {}
    elif current_section and first_cell == "( Add into cetralised destination Block)":
        # row[1] holds the integer to add
        De_Centralized_data[current_section]["Add"] = int(row[1]) if pd.notna(row[1]) else 0

# Ensure keys exist for both "Semi Centralized" and "DeCentralised"
# If missing, default Add = 0
for key in ["Semi Centralized", "DeCentralised"]:
    if key not in De_Centralized_data:
        De_Centralized_data[key] = {"Add": 0}
    elif "Add" not in De_Centralized_data[key]:
        De_Centralized_data[key]["Add"] = 0

# ----------------------------------------
# Step 2: Preprocess Blocks & Department Split
# ----------------------------------------

# 2.1 Convert cumulative circulation area from SQFT → SQM
all_block_data['Cumulative_Area_SQM'] = (
    all_block_data['Cumulative_Block_Circulation_Area_(SQFT)'] / 10.764
)

# 2.2 Separate Destination vs. Typical blocks
destination_blocks = all_block_data[
    all_block_data['Typical_Destination'] == 'Destination'
].copy()
typical_blocks = all_block_data[
    all_block_data['Typical_Destination'] == 'Typical'
].copy()

# 2.3 Dept‐split dict (not needed for new logic but loaded)
dept_split_dict = department_split_data.set_index('Department_Sub-Department')['Splittable'].to_dict()

# ----------------------------------------
# Step 3: Initialize Floor Assignments
# ----------------------------------------

def initialize_floor_assignments(floor_df):
    """
    Returns a dict keyed by floor name. Each entry contains:
      - remaining_area
      - remaining_capacity
      - assigned_blocks      (list of block‐row dicts)
      - assigned_departments (set of sub‐departments)
      - ME_area, WE_area, US_area, Support_area, Speciality_area (floats)
    """
    assignments = {}
    for _, row in floor_df.iterrows():
        floor = row['Name'].strip()
        assignments[floor] = {
            'remaining_area': row['Usable_Area_(SQM)'],
            'remaining_capacity': row['Max_Assignable_Floor_loading_Capacity'],
            'assigned_blocks': [],
            'assigned_departments': set(),
            'ME_area': 0.0,
            'WE_area': 0.0,
            'US_area': 0.0,
            'Support_area': 0.0,
            'Speciality_area': 0.0
        }
    return assignments

floors = list(initialize_floor_assignments(all_floor_data).keys())

# ----------------------------------------
# Step 4: Core Stacking Function
# ----------------------------------------

def run_stack_plan(mode):
    """
    mode: 'centralized', 'semi', or 'decentralized'
    Returns:
      1) detailed_df      – each block’s assigned floor, department, block name,
                            destination group, space mix, area (SQM), occupancy
      2) floor_summary_df – for each floor: total number of blocks, total area, total occupancy
      3) space_mix_df     – for each floor and each category {ME, WE, US, Support, Speciality}:
                              - Unit_Count_on_Floor
                              - Pct_of_Floor_UC      (floor count / total blocks on that floor ×100%)
                              - Pct_of_Overall_UC    (floor count / total blocks of that category ×100%)
    """
    assignments = initialize_floor_assignments(all_floor_data)

    def floors_allowed_for_dest_in_group():
        """
        Determine how many floors to use for destination blocks, based on mode:
          - 'centralized': exactly 2
          - 'semi':       2 + De_Centralized_data["Semi Centralized"]["Add"]
          - 'decentralized': 2 + De_Centralized_data["DeCentralised"]["Add"]
        """
        if mode == 'centralized':
            return 2
        elif mode == 'semi':
            return 2 + De_Centralized_data.get("Semi Centralized", {}).get("Add", 0)
        elif mode == 'decentralized':
            return 2 + De_Centralized_data.get("DeCentralised", {}).get("Add", 0)
        else:
            return 2

    max_dest_floors = floors_allowed_for_dest_in_group()

    # 4.1 Assign Destination Blocks (largest‐area first within each Destination_Group)
    for group_name, group_df in destination_blocks.groupby('Destination_Group'):
        sorted_blocks = group_df.sort_values(by='Cumulative_Area_SQM', ascending=False).to_dict('records')
        target_floors = floors[:max_dest_floors]

        for blk in sorted_blocks:
            blk_area     = blk['Cumulative_Area_SQM']
            blk_capacity = blk['Max_Occupancy_with_Capacity']
            blk_dept     = blk['Department_Sub_Department'].strip()
            placed = False

            # Try only those target destination floors first
            for fl in target_floors:
                if assignments[fl]['remaining_area'] >= blk_area:
                    assignments[fl]['assigned_blocks'].append(blk)
                    assignments[fl]['remaining_area'] -= blk_area
                    assignments[fl]['remaining_capacity'] -= blk_capacity
                    assignments[fl]['assigned_departments'].add(blk_dept)
                    placed = True
                    break

            if not placed:
                # If none of those fit, keep going upward in the building
                for fl in floors[max_dest_floors:]:
                    if assignments[fl]['remaining_area'] >= blk_area:
                        assignments[fl]['assigned_blocks'].append(blk)
                        assignments[fl]['remaining_area'] -= blk_area
                        assignments[fl]['remaining_capacity'] -= blk_capacity
                        assignments[fl]['assigned_departments'].add(blk_dept)
                        placed = True
                        break

            if not placed:
                print(f"Warning: Could not place destination block '{blk['Block_Name']}'")

    # 4.2 Phase: Randomly assign ME blocks first (typical)
    me_blocks = [
        blk for blk in typical_blocks.to_dict('records')
        if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == 'ME'
    ]
    random.shuffle(me_blocks)
    for blk in me_blocks:
        blk_area     = blk['Cumulative_Area_SQM']
        blk_capacity = blk['Max_Occupancy_with_Capacity']
        blk_dept     = blk['Department_Sub_Department'].strip()

        candidate_floors = floors.copy()
        random.shuffle(candidate_floors)
        placed = False
        for fl in candidate_floors:
            if assignments[fl]['remaining_area'] >= blk_area:
                assignments[fl]['assigned_blocks'].append(blk)
                assignments[fl]['remaining_area'] -= blk_area
                assignments[fl]['remaining_capacity'] -= blk_capacity
                assignments[fl]['assigned_departments'].add(blk_dept)
                assignments[fl]['ME_area'] += blk_area
                placed = True
                break
        if not placed:
            print(f"Warning: Could not place ME block '{blk['Block_Name']}'")

    # Compute ME distribution per floor (count)
    me_count_per_floor = {fl: 0 for fl in floors}
    for fl, info in assignments.items():
        me_count_per_floor[fl] = sum(
            1 for blk in info['assigned_blocks']
            if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == 'ME'
        )
    total_me = sum(me_count_per_floor.values())
    if total_me == 0:
        # If zero ME blocks, distribute fractions equally across floors
        me_frac_per_floor = {fl: 1 / len(floors) for fl in floors}
    else:
        me_frac_per_floor = {fl: me_count_per_floor[fl] / total_me for fl in floors}

    # 4.3 Phase: Assign other categories proportionally
    other_categories = ['WE', 'US', 'Support', 'Speciality']
    for category in other_categories:
        cat_blocks = [
            blk for blk in typical_blocks.to_dict('records')
            if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == category
        ]
        total_cat = len(cat_blocks)
        if total_cat == 0:
            continue

        # Raw target per floor = me_frac_per_floor * total_cat
        raw_targets = {fl: me_frac_per_floor[fl] * total_cat for fl in floors}
        target_counts = {fl: int(round(raw_targets[fl])) for fl in floors}

        # Adjust rounding so that sum(target_counts) == total_cat
        diff = total_cat - sum(target_counts.values())
        if diff != 0:
            fractional_parts = {
                fl: raw_targets[fl] - math.floor(raw_targets[fl]) for fl in floors
            }
            if diff > 0:
                # Add 1 to floors with largest fractional parts
                for fl in sorted(floors, key=lambda x: fractional_parts[x], reverse=True)[:diff]:
                    target_counts[fl] += 1
            else:
                # Remove 1 from floors with smallest fractional parts
                for fl in sorted(floors, key=lambda x: fractional_parts[x])[: -diff]:
                    target_counts[fl] -= 1

        random.shuffle(cat_blocks)
        assigned_counts = {fl: 0 for fl in floors}

        for blk in cat_blocks:
            blk_area     = blk['Cumulative_Area_SQM']
            blk_capacity = blk['Max_Occupancy_with_Capacity']
            blk_dept     = blk['Department_Sub_Department'].strip()

            # Floors with a positive “deficit” first
            deficits = {fl: target_counts[fl] - assigned_counts[fl] for fl in floors}
            floors_with_deficit = [fl for fl, d in deficits.items() if d > 0]
            if floors_with_deficit:
                candidate_floors = sorted(
                    floors_with_deficit,
                    key=lambda x: deficits[x],
                    reverse=True
                )
            else:
                candidate_floors = floors.copy()

            placed = False
            for fl in candidate_floors:
                if assignments[fl]['remaining_area'] >= blk_area:
                    assignments[fl]['assigned_blocks'].append(blk)
                    assignments[fl]['remaining_area'] -= blk_area
                    assignments[fl]['remaining_capacity'] -= blk_capacity
                    assignments[fl]['assigned_departments'].add(blk_dept)
                    if category == 'WE':
                        assignments[fl]['WE_area'] += blk_area
                    elif category == 'US':
                        assignments[fl]['US_area'] += blk_area
                    elif category == 'Support':
                        assignments[fl]['Support_area'] += blk_area
                    elif category == 'Speciality':
                        assignments[fl]['Speciality_area'] += blk_area
                    assigned_counts[fl] += 1
                    placed = True
                    break

            if not placed:
                # Fallback: any floor in random order
                fallback = floors.copy()
                random.shuffle(fallback)
                for fl in fallback:
                    if assignments[fl]['remaining_area'] >= blk_area:
                        assignments[fl]['assigned_blocks'].append(blk)
                        assignments[fl]['remaining_area'] -= blk_area
                        assignments[fl]['remaining_capacity'] -= blk_capacity
                        assignments[fl]['assigned_departments'].add(blk_dept)
                        if category == 'WE':
                            assignments[fl]['WE_area'] += blk_area
                        elif category == 'US':
                            assignments[fl]['US_area'] += blk_area
                        elif category == 'Support':
                            assignments[fl]['Support_area'] += blk_area
                        elif category == 'Speciality':
                            assignments[fl]['Speciality_area'] += blk_area
                        assigned_counts[fl] += 1
                        placed = True
                        break

            if not placed:
                print(f"Warning: Could not place {category} block '{blk['Block_Name']}'")

    # Phase 3: Build Detailed & Summary DataFrames

    # 3.1 “Detailed” DataFrame
    assignment_list = []
    for fl, info in assignments.items():
        for blk in info['assigned_blocks']:
            assignment_list.append({
                'Floor': fl,
                'Department': blk['Department_Sub_Department'],
                'Block_Name': blk['Block_Name'],
                'Destination_Group': blk['Destination_Group'],
                'SpaceMix': blk['SpaceMix_(ME_WE_US_Support_Speciality)'],
                'Assigned_Area_SQM': blk['Cumulative_Area_SQM'],
                'Max_Occupancy': blk['Max_Occupancy_with_Capacity']
            })
    detailed_df = pd.DataFrame(assignment_list)

    # 3.2 “Floor_Summary” DataFrame
    floor_summary_df = (
        detailed_df
        .groupby('Floor')
        .agg(
            Total_Blocks=('Block_Name', 'count'),
            Total_Area_SQM=('Assigned_Area_SQM', 'sum'),
            Total_Occupancy=('Max_Occupancy', 'sum')
        )
        .reset_index()
    )

    # 3.3 “SpaceMix_By_Units” DataFrame
    all_categories = ['ME', 'WE', 'US', 'Support', 'Speciality']
    category_totals = {
        cat: len(typical_blocks[
            typical_blocks['SpaceMix_(ME_WE_US_Support_Speciality)'].str.strip() == cat
        ])
        for cat in all_categories
    }

    rows = []
    for fl, info in assignments.items():
        counts = {cat: 0 for cat in all_categories}
        for blk in info['assigned_blocks']:
            cat = blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip()
            if cat in counts:
                counts[cat] += 1
        total_blocks_on_floor = sum(counts.values())

        for cat in all_categories:
            cnt = counts[cat]
            pct_of_floor = (cnt / total_blocks_on_floor * 100) if total_blocks_on_floor else 0.0
            total_cat = category_totals[cat]
            pct_overall = (cnt / total_cat * 100) if total_cat else 0.0

            rows.append({
                'Floor': fl,
                'SpaceMix': cat,
                'Unit_Count_on_Floor': cnt,
                'Pct_of_Floor_UC': round(pct_of_floor, 2),
                'Pct_of_Overall_UC': round(pct_overall, 2)
            })

    space_mix_df = pd.DataFrame(rows)

    return detailed_df, floor_summary_df, space_mix_df

# ----------------------------------------
# Step 5: Generate & Export Three Excel Files
# ----------------------------------------

# 5.1 Centralized (strictly 2 floors for destinations)
central_detailed, central_floor_sum, central_space_mix = run_stack_plan('centralized')

# 5.2 Semi Centralized (2 + “Add” from sheet)
semi_detailed, semi_floor_sum, semi_space_mix = run_stack_plan('semi')

# 5.3 Decentralized (2 + “Add” from sheet)
decentral_detailed, decentral_floor_sum, decentral_space_mix = run_stack_plan('decentralized')

# Write each to its own Excel file
with pd.ExcelWriter('stack_plan_centralized5.xlsx') as writer:
    central_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    central_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    central_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

with pd.ExcelWriter('stack_plan_semi_centralized5.xlsx') as writer:
    semi_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    semi_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    semi_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

with pd.ExcelWriter('stack_plan_decentralized5.xlsx') as writer:
    decentral_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    decentral_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    decentral_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

print("✅ Generated three Excel outputs:")
print("    • stack_plan_centralized5.xlsx")
print("    • stack_plan_semi_centralized5.xlsx")
print("    • stack_plan_decentralized5.xlsx")
```

```python
import pandas as pd
import random
import math

# ----------------------------------------
# Step 1: Load Input Sheets
# ----------------------------------------

excel_path = '/content/AR-1.xlsx'  # ← adjust if needed

# 1.1 Floors sheet
all_floor_data = pd.read_excel(excel_path, sheet_name='Program Table Input 2 - Floor')
all_floor_data.columns = all_floor_data.columns.str.strip()

# 1.2 Blocks sheet
all_block_data = pd.read_excel(excel_path, sheet_name='Program Table Input 1 - Block')
all_block_data.columns = all_block_data.columns.str.strip()

# 1.3 Department Split sheet
department_split_data = pd.read_excel(excel_path, sheet_name='Department Split', skiprows=1)
department_split_data.columns = department_split_data.columns.str.strip()
department_split_data = department_split_data.rename(
    columns={'BU_Department_Sub-Department': 'Department_Sub-Department'}
)

# 1.4 Adjacency sheet
xls = pd.ExcelFile(excel_path)
adjacency_sheet_name = [name for name in xls.sheet_names if "Adjacency" in name][0]
raw_data = xls.parse(adjacency_sheet_name, header=1, index_col=0)
adjacency_data = raw_data.apply(pd.to_numeric, errors='coerce')
adjacency_data.index = adjacency_data.index.str.strip()
adjacency_data.columns = adjacency_data.columns.str.strip()

# 1.5 De-Centralized Logic sheet
# We expect rows: "Centralised", "Semi Centralized", "DeCentralised" in column 0,
# and their "( Add into cetralised destination Block)" values in column 1.
df_logic = pd.read_excel(excel_path, sheet_name='De-Centralized Logic', header=None)
De_Centralized_data = {}
current_section = None
for _, row in df_logic.iterrows():
    first_cell = str(row[0]).strip() if pd.notna(row[0]) else ""
    if first_cell in ["Centralised", "Semi Centralized", "DeCentralised"]:
        current_section = first_cell
        De_Centralized_data[current_section] = {"Add": 0}
    elif current_section and first_cell == "( Add into cetralised destination Block)":
        De_Centralized_data[current_section]["Add"] = int(row[1]) if pd.notna(row[1]) else 0

# Ensure keys exist
for key in ["Centralised", "Semi Centralized", "DeCentralised"]:
    if key not in De_Centralized_data:
        De_Centralized_data[key] = {"Add": 0}
    elif "Add" not in De_Centralized_data[key]:
        De_Centralized_data[key]["Add"] = 0

# ----------------------------------------
# Step 2: Preprocess Blocks & Department Split
# ----------------------------------------

# 2.1 Convert cumulative circulation area from SQFT → SQM
all_block_data['Cumulative_Area_SQM'] = (
    all_block_data['Cumulative_Block_Circulation_Area_(SQFT)'] / 10.764
)

# 2.2 Separate Destination vs. Typical blocks
destination_blocks = all_block_data[
    all_block_data['Typical_Destination'] == 'Destination'
].copy()
typical_blocks = all_block_data[
    all_block_data['Typical_Destination'] == 'Typical'
].copy()

# 2.3 Dept‐split dict (not used in destination logic but loaded for completeness)
dept_split_dict = department_split_data.set_index('Department_Sub-Department')['Splittable'].to_dict()

# ----------------------------------------
# Step 3: Initialize Floor Assignments
# ----------------------------------------

def initialize_floor_assignments(floor_df):
    """
    Returns a dict keyed by floor name. Each entry tracks:
      - remaining_area
      - remaining_capacity
      - assigned_blocks      (list of block‐row dicts)
      - assigned_departments (set of sub‐departments)
      - ME_area, WE_area, US_area, Support_area, Speciality_area (floats)
    """
    assignments = {}
    for _, row in floor_df.iterrows():
        floor = row['Name'].strip()
        assignments[floor] = {
            'remaining_area': row['Usable_Area_(SQM)'],
            'remaining_capacity': row['Max_Assignable_Floor_loading_Capacity'],
            'assigned_blocks': [],
            'assigned_departments': set(),
            'ME_area': 0.0,
            'WE_area': 0.0,
            'US_area': 0.0,
            'Support_area': 0.0,
            'Speciality_area': 0.0
        }
    return assignments

floors = list(initialize_floor_assignments(all_floor_data).keys())

# ----------------------------------------
# Step 4: Core Stacking Function
# ----------------------------------------

def run_stack_plan(mode):
    """
    mode: 'centralized', 'semi', or 'decentralized'
    Returns three DataFrames:
      1) detailed_df      – each block’s assigned floor, department, block name, destination group, space mix, area, occupancy
      2) floor_summary_df – floor‐wise totals (block count, total area, total occupancy)
      3) space_mix_df     – for each floor and each category {ME, WE, US, Support, Speciality}:
                              - Unit_Count_on_Floor
                              - Pct_of_Floor_UC      = (category_count_on_floor / total_blocks_on_floor) × 100%
                              - Pct_of_Overall_UC    = (category_count_on_floor / total_blocks_of_category_overall) × 100%
    """
    assignments = initialize_floor_assignments(all_floor_data)

    # Determine how many floors to use for destination blocks
    def destination_floor_count():
        if mode == 'centralized':
            return 2
        elif mode == 'semi':
            return 2 + De_Centralized_data["Semi Centralized"]["Add"]
        elif mode == 'decentralized':
            return 2 + De_Centralized_data["DeCentralised"]["Add"]
        else:
            return 2

    max_dest_floors = destination_floor_count()
    # Cap at total number of floors
    max_dest_floors = min(max_dest_floors, len(floors))

    # Pre‐compute each group's total area and total capacity
    dest_groups = {}
    for _, blk in destination_blocks.iterrows():
        grp = blk['Destination_Group']
        if grp not in dest_groups:
            dest_groups[grp] = {'blocks': [], 'total_area': 0.0, 'total_capacity': 0}
        dest_groups[grp]['blocks'].append(blk.to_dict())
        dest_groups[grp]['total_area'] += blk['Cumulative_Area_SQM']
        dest_groups[grp]['total_capacity'] += blk['Max_Occupancy_with_Capacity']

    # Phase 1: Assign entire destination groups randomly among the first max_dest_floors
    group_names = list(dest_groups.keys())
    random.shuffle(group_names)

    for grp in group_names:
        group_info = dest_groups[grp]
        grp_area = group_info['total_area']
        grp_cap  = group_info['total_capacity']
        placed = False

        # Shuffle allowed floors among the first max_dest_floors
        candidate_floors = floors[:max_dest_floors].copy()
        random.shuffle(candidate_floors)

        for fl in candidate_floors:
            if (assignments[fl]['remaining_area'] >= grp_area and
                assignments[fl]['remaining_capacity'] >= grp_cap):
                # Assign entire group to that floor
                for blk in group_info['blocks']:
                    assignments[fl]['assigned_blocks'].append(blk)
                    assignments[fl]['assigned_departments'].add(blk['Department_Sub_Department'].strip())
                assignments[fl]['remaining_area'] -= grp_area
                assignments[fl]['remaining_capacity'] -= grp_cap
                placed = True
                break

        if not placed:
            # If no allowed floor fits, try any floor above max_dest_floors
            for fl in floors[max_dest_floors:]:
                if (assignments[fl]['remaining_area'] >= grp_area and
                    assignments[fl]['remaining_capacity'] >= grp_cap):
                    for blk in group_info['blocks']:
                        assignments[fl]['assigned_blocks'].append(blk)
                        assignments[fl]['assigned_departments'].add(blk['Department_Sub_Department'].strip())
                    assignments[fl]['remaining_area'] -= grp_area
                    assignments[fl]['remaining_capacity'] -= grp_cap
                    placed = True
                    break

        if not placed:
            print(f"Warning: Could not place entire destination group '{grp}'")

    # Phase 2A: Randomly assign ME blocks (typical)
    me_blocks = [
        blk for blk in typical_blocks.to_dict('records')
        if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == 'ME'
    ]
    random.shuffle(me_blocks)
    for blk in me_blocks:
        blk_area     = blk['Cumulative_Area_SQM']
        blk_capacity = blk['Max_Occupancy_with_Capacity']
        blk_dept     = blk['Department_Sub_Department'].strip()

        candidate_floors = floors.copy()
        random.shuffle(candidate_floors)
        placed = False
        for fl in candidate_floors:
            if assignments[fl]['remaining_area'] >= blk_area:
                assignments[fl]['assigned_blocks'].append(blk)
                assignments[fl]['remaining_area'] -= blk_area
                assignments[fl]['remaining_capacity'] -= blk_capacity
                assignments[fl]['assigned_departments'].add(blk_dept)
                assignments[fl]['ME_area'] += blk_area
                placed = True
                break
        if not placed:
            print(f"Warning: Could not place ME block '{blk['Block_Name']}'")

    # Compute ME distribution per floor (unit counts)
    me_count_per_floor = {fl: 0 for fl in floors}
    for fl, info in assignments.items():
        me_count_per_floor[fl] = sum(
            1 for blk in info['assigned_blocks']
            if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == 'ME'
        )
    total_me = sum(me_count_per_floor.values())
    if total_me == 0:
        me_frac_per_floor = {fl: 1 / len(floors) for fl in floors}
    else:
        me_frac_per_floor = {fl: me_count_per_floor[fl] / total_me for fl in floors}

    # Phase 2B: Assign other categories proportionally
    other_categories = ['WE', 'US', 'Support', 'Speciality']
    for category in other_categories:
        cat_blocks = [
            blk for blk in typical_blocks.to_dict('records')
            if blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip() == category
        ]
        total_cat = len(cat_blocks)
        if total_cat == 0:
            continue

        raw_targets = {fl: me_frac_per_floor[fl] * total_cat for fl in floors}
        target_counts = {fl: int(round(raw_targets[fl])) for fl in floors}

        diff = total_cat - sum(target_counts.values())
        if diff != 0:
            fractional_parts = {
                fl: raw_targets[fl] - math.floor(raw_targets[fl]) for fl in floors
            }
            if diff > 0:
                for fl in sorted(floors, key=lambda x: fractional_parts[x], reverse=True)[:diff]:
                    target_counts[fl] += 1
            else:
                for fl in sorted(floors, key=lambda x: fractional_parts[x])[: -diff]:
                    target_counts[fl] -= 1

        random.shuffle(cat_blocks)
        assigned_counts = {fl: 0 for fl in floors}

        for blk in cat_blocks:
            blk_area     = blk['Cumulative_Area_SQM']
            blk_capacity = blk['Max_Occupancy_with_Capacity']
            blk_dept     = blk['Department_Sub_Department'].strip()

            deficits = {fl: target_counts[fl] - assigned_counts[fl] for fl in floors}
            floors_with_deficit = [fl for fl, d in deficits.items() if d > 0]
            if floors_with_deficit:
                candidate_floors = sorted(
                    floors_with_deficit,
                    key=lambda x: deficits[x],
                    reverse=True
                )
            else:
                candidate_floors = floors.copy()

            placed = False
            for fl in candidate_floors:
                if assignments[fl]['remaining_area'] >= blk_area:
                    assignments[fl]['assigned_blocks'].append(blk)
                    assignments[fl]['remaining_area'] -= blk_area
                    assignments[fl]['remaining_capacity'] -= blk_capacity
                    assignments[fl]['assigned_departments'].add(blk_dept)
                    if category == 'WE':
                        assignments[fl]['WE_area'] += blk_area
                    elif category == 'US':
                        assignments[fl]['US_area'] += blk_area
                    elif category == 'Support':
                        assignments[fl]['Support_area'] += blk_area
                    elif category == 'Speciality':
                        assignments[fl]['Speciality_area'] += blk_area
                    assigned_counts[fl] += 1
                    placed = True
                    break

            if not placed:
                fallback = floors.copy()
                random.shuffle(fallback)
                for fl in fallback:
                    if assignments[fl]['remaining_area'] >= blk_area:
                        assignments[fl]['assigned_blocks'].append(blk)
                        assignments[fl]['remaining_area'] -= blk_area
                        assignments[fl]['remaining_capacity'] -= blk_capacity
                        assignments[fl]['assigned_departments'].add(blk_dept)
                        if category == 'WE':
                            assignments[fl]['WE_area'] += blk_area
                        elif category == 'US':
                            assignments[fl]['US_area'] += blk_area
                        elif category == 'Support':
                            assignments[fl]['Support_area'] += blk_area
                        elif category == 'Speciality':
                            assignments[fl]['Speciality_area'] += blk_area
                        assigned_counts[fl] += 1
                        placed = True
                        break

            if not placed:
                print(f"Warning: Could not place {category} block '{blk['Block_Name']}'")

    # Phase 3: Build Detailed & Summary DataFrames

    # 3.1 Detailed DataFrame
    assignment_list = []
    for fl, info in assignments.items():
        for blk in info['assigned_blocks']:
            assignment_list.append({
                'Floor': fl,
                'Department': blk['Department_Sub_Department'],
                'Block_Name': blk['Block_Name'],
                'Destination_Group': blk['Destination_Group'],
                'SpaceMix': blk['SpaceMix_(ME_WE_US_Support_Speciality)'],
                'Assigned_Area_SQM': blk['Cumulative_Area_SQM'],
                'Max_Occupancy': blk['Max_Occupancy_with_Capacity']
            })
    detailed_df = pd.DataFrame(assignment_list)

    # 3.2 Floor_Summary DataFrame
    floor_summary_df = (
        detailed_df
        .groupby('Floor')
        .agg(
            Total_Blocks=('Block_Name', 'count'),
            Total_Area_SQM=('Assigned_Area_SQM', 'sum'),
            Total_Occupancy=('Max_Occupancy', 'sum')
        )
        .reset_index()
    )

    # 3.3 SpaceMix_By_Units DataFrame
    all_categories = ['ME', 'WE', 'US', 'Support', 'Speciality']
    category_totals = {
        cat: len(typical_blocks[
            typical_blocks['SpaceMix_(ME_WE_US_Support_Speciality)'].str.strip() == cat
        ])
        for cat in all_categories
    }

    rows = []
    for fl, info in assignments.items():
        counts = {cat: 0 for cat in all_categories}
        for blk in info['assigned_blocks']:
            cat = blk['SpaceMix_(ME_WE_US_Support_Speciality)'].strip()
            if cat in counts:
                counts[cat] += 1
        total_blocks_on_floor = sum(counts.values())

        for cat in all_categories:
            cnt = counts[cat]
            pct_of_floor = (cnt / total_blocks_on_floor * 100) if total_blocks_on_floor else 0.0
            total_cat = category_totals[cat]
            pct_overall = (cnt / total_cat * 100) if total_cat else 0.0

            rows.append({
                'Floor': fl,
                'SpaceMix': cat,
                'Unit_Count_on_Floor': cnt,
                'Pct_of_Floor_UC': round(pct_of_floor, 2),
                'Pct_of_Overall_UC': round(pct_overall, 2)
            })

    space_mix_df = pd.DataFrame(rows)

    return detailed_df, floor_summary_df, space_mix_df

# ----------------------------------------
# Step 5: Generate & Export Three Excel Files
# ----------------------------------------

central_detailed, central_floor_sum, central_space_mix = run_stack_plan('centralized')
semi_detailed, semi_floor_sum, semi_space_mix = run_stack_plan('semi')
decentral_detailed, decentral_floor_sum, decentral_space_mix = run_stack_plan('decentralized')

with pd.ExcelWriter('stack_plan_centralized6.xlsx') as writer:
    central_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    central_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    central_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

with pd.ExcelWriter('stack_plan_semi_centralized6.xlsx') as writer:
    semi_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    semi_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    semi_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

with pd.ExcelWriter('stack_plan_decentralized6.xlsx') as writer:
    decentral_detailed.to_excel(writer, sheet_name='Detailed', index=False)
    decentral_floor_sum.to_excel(writer, sheet_name='Floor_Summary', index=False)
    decentral_space_mix.to_excel(writer, sheet_name='SpaceMix_By_Units', index=False)

print("✅ Generated three Excel outputs:")
print("    • stack_plan_centralized6.xlsx")
print("    • stack_plan_semi_centralized6.xlsx")
print("    • stack_plan_decentralized6.xlsx")
```
