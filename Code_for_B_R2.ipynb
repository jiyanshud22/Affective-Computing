```python
import pandas as pd
import numpy as np
import networkx as nx
from itertools import combinations
from random import choice, seed

# ————— 1) Load inputs —————
adj = pd.read_csv("/content/BR-2_adjacency.csv", index_col=0)
blocks = all_block_data.copy()   # must have Department_Sub-Department & Area

# list of floors
floors = ["Floor 1","Floor 2","Floor 3","Floor 4"]

# For reproducibility
seed(42)
np.random.seed(42)

# ————— 2) Extract pair-lists by score —————
mask = np.triu(np.ones(adj.shape, dtype=bool), k=1)
long = (
    adj.where(mask)
       .stack()
       .reset_index()
       .rename(columns={"level_0":"D1","level_1":"D2",0:"score"})
)

hard_same = long[ long["score"] ==  1.0 ]   # must be same floor
hard_diff = long[ long["score"] == -1.0 ]   # must be different floors
soft     = long[ long["score"].isin({0.25, 0.5, 0.75}) ]
zero     = long[ long["score"] ==  0.0 ]

# ————— 3) Build “must-co-locate” clusters from score=1 edges —————
G = nx.Graph()
G.add_nodes_from(adj.index)
for _,r in hard_same.iterrows():
    G.add_edge(r["D1"], r["D2"])

clusters = list(nx.connected_components(G))
# map each sub-dept → its cluster id
cluster_of = { d: cid for cid,comp in enumerate(clusters) for d in comp }
blocks["cluster"] = blocks["Department_Sub-Department"].map(cluster_of)

# precompute all cluster→cluster soft-weights
cluster_weight = {}
for (i,j,w) in soft.itertuples(index=False, name=None):
    c1, c2 = cluster_of[i], cluster_of[j]
    if c1==c2:
        continue
    key = tuple(sorted((c1,c2)))
    cluster_weight[key] = cluster_weight.get(key, 0) + w

# precompute cluster pairs that must be different floors
neg_pairs = set()
for _,r in hard_diff.iterrows():
    c1, c2 = cluster_of[r["D1"]], cluster_of[r["D2"]]
    if c1!=c2:
        neg_pairs.add(tuple(sorted((c1,c2))))

# ————— 4) Greedy cluster→floor assignment —————
cluster_to_floor = {}
assigned = []

for cid in range(len(clusters)):
    # which floors are still legal (no -1 conflict)
    legal = []
    for f in floors:
        conflict = False
        for oc in assigned:
            if cluster_to_floor[oc] == f and tuple(sorted((cid,oc))) in neg_pairs:
                conflict = True
                break
        if not conflict:
            legal.append(f)
    if not legal:
        # if absolutely no legal floor, fall back to all floors
        legal = floors[:]
    # pick the floor that maximizes soft-weight connections with already placed clusters
    best, best_score = None, -1e9
    for f in legal:
        score = sum(
            cluster_weight.get(tuple(sorted((cid,oc))),0)
            for oc in assigned
            if cluster_to_floor[oc] == f
        )
        if score > best_score:
            best_score, best = score, f
    # tie-break or first cluster
    if best is None:
        best = choice(legal)
    cluster_to_floor[cid] = best
    assigned.append(cid)

# tag every block by its cluster floor
blocks["floor"] = blocks["cluster"].map(cluster_to_floor)

# ————— 5) Fill in the zero-adjacency blocks to match overall mix —————
# compute total area & per-dept global ratio
total_area = blocks["Area"].sum()
dept_total = blocks.groupby("Department_Sub-Department")["Area"].sum()
global_ratio = dept_total / total_area

# assume equal floor capacity = total_area / #floors
floor_cap = { f: total_area/len(floors) for f in floors }

# track how much area is used on each floor, and per-dept on each floor
used_area = { f: blocks[blocks["floor"]==f]["Area"].sum() for f in floors }
dept_used = {
    f: blocks[blocks["floor"]==f]
          .groupby("Department_Sub-Department")["Area"]
          .sum()
          .to_dict()
    for f in floors
}

# get all blocks whose sub-dept never appeared in any non-zero pair (i.e. score=0 in long)
zero_subs = set(zero["D1"]) | set(zero["D2"])
zero_blocks = blocks[blocks["Department_Sub-Department"].isin(zero_subs)].copy()

# sort largest first
zero_blocks = zero_blocks.sort_values("Area", ascending=False)

for idx,blk in zero_blocks.iterrows():
    d = blk["Department_Sub-Department"]
    a = blk["Area"]
    # pick the floor where this dept is *most under-represented* vs global ratio
    best_f, best_dev = None, 1e9
    for f in floors:
        rem = floor_cap[f] - used_area[f]
        if a > rem:
            continue
        cur_ratio = dept_used[f].get(d,0) / floor_cap[f]
        dev = abs(cur_ratio - global_ratio[d])
        if dev < best_dev:
            best_dev, best_f = dev, f

    if best_f is None:
        # if nothing fits, dump onto the floor with the most free space
        best_f = max(floors, key=lambda f: floor_cap[f] - used_area[f])

    # assign
    blocks.at[idx, "floor"] = best_f
    used_area[best_f] += a
    dept_used[best_f][d] = dept_used[best_f].get(d,0) + a

# ————— 6) (Optional) Export your final assignment —————
out = blocks[["Department_Sub-Department","Area","floor"]].rename(
    columns={"Department_Sub-Department":"SubDept",
             "floor":"AssignedFloor"}
)
out.to_excel("/content/final_floor_assignments.xlsx", index=False, sheet_name="Assignments")
print("Done → /content/final_floor_assignments.xlsx")
```

```python
import pandas as pd
import numpy as np
import networkx as nx
from random import choice, seed

# ——— 1) LOAD INPUTS —————————————————————————————
# 1.a) adjacency matrix (ensure floats)
adj = pd.read_csv("/content/BR-2_adjacency.csv", index_col=0)
adj = adj.astype(float).round(2)

# 1.b) your blocks table (must have Dept key + Block Area)
#      e.g.: all_block_data = pd.read_csv("/content/all_block_data.csv")
blocks = all_block_data.copy()

# rename the area column so we can refer to blocks["Area"]
blocks.rename(columns={"Block Area  (SQFT)": "Area"}, inplace=True)

# ——— 2) SETUP FLOORS + RNG SEEDS ——————————————————————
floors = ["Floor 1", "Floor 2", "Floor 3", "Floor 4"]
seed(42)
np.random.seed(42)

# ——— 3) EXTRACT ALL PAIRS + SPLIT BY SCORE —————————
mask = np.triu(np.ones(adj.shape, dtype=bool), k=1)
pairs = (
    adj.where(mask)
       .stack()
       .reset_index()
       .rename(columns={
           "level_0": "D1",
           "level_1": "D2",
           0:         "score"
       })
)

hard_same = pairs[np.isclose(pairs["score"],  1.00)]   # must co‐locate
hard_diff = pairs[np.isclose(pairs["score"], -1.00)]   # must separate
soft      = pairs[pairs["score"].isin({0.25, 0.50, 0.75})]
zero      = pairs[np.isclose(pairs["score"],  0.00)]

# ——— 4) BUILD “MUST‐COLLOCATE” CLUSTERS (score=1) —————
G = nx.Graph()
G.add_nodes_from(adj.index)
for _, r in hard_same.iterrows():
    G.add_edge(r["D1"], r["D2"])

clusters = list(nx.connected_components(G))
cluster_of = {
    dept: cid
    for cid, comp in enumerate(clusters)
    for dept in comp
}
blocks["cluster"] = blocks["Department_Sub-Department"].map(cluster_of)

# ——— 5) PRECOMPUTE SOFT WEIGHTS & “MUST‐SEPARATE” PAIRS ————
# soft‐weights per cluster‐pair
cluster_weight = {}
for d1, d2, w in soft.itertuples(index=False, name=None):
    c1, c2 = cluster_of[d1], cluster_of[d2]
    if c1 == c2:
        continue
    key = tuple(sorted((c1, c2)))
    cluster_weight[key] = cluster_weight.get(key, 0) + w

# pairs of clusters that must be on different floors
must_diff = {
    tuple(sorted((cluster_of[r["D1"]], cluster_of[r["D2"]])))
    for _, r in hard_diff.iterrows()
    if cluster_of[r["D1"]] != cluster_of[r["D2"]]
}

# ——— 6) GREEDY ASSIGNMENT OF CLUSTERS TO FLOORS ————————
cluster_to_floor = {}
placed = []

for cid in range(len(clusters)):
    # 6.a) legal floors for this cluster (no must-diff violation)
    legal = []
    for f in floors:
        bad = False
        for oc in placed:
            if cluster_to_floor[oc] == f and tuple(sorted((cid, oc))) in must_diff:
                bad = True
                break
        if not bad:
            legal.append(f)
    if not legal:
        legal = floors[:]  # fallback to any

    # 6.b) pick floor maximizing sum of soft‐weights with already‐placed clusters
    best_floor, best_score = None, -1e9
    for f in legal:
        score = sum(
            cluster_weight.get(tuple(sorted((cid, oc))), 0)
            for oc in placed
            if cluster_to_floor[oc] == f
        )
        if score > best_score:
            best_score, best_floor = score, f

    if best_floor is None:
        best_floor = choice(legal)
    cluster_to_floor[cid] = best_floor
    placed.append(cid)

# map back to blocks
blocks["floor"] = blocks["cluster"].map(cluster_to_floor)

# ——— 7) PLACE ZERO‐ADJ BLOCKS TO MATCH GLOBAL MIX —————
total_area = blocks["Area"].sum()
dept_totals = blocks.groupby("Department_Sub-Department")["Area"].sum()
global_ratio = dept_totals / total_area
floor_capacity = {f: total_area/len(floors) for f in floors}

used_area = {
    f: blocks.loc[blocks["floor"]==f, "Area"].sum()
    for f in floors
}
dept_used = {
    f: blocks[blocks["floor"]==f]
           .groupby("Department_Sub-Department")["Area"]
           .sum()
           .to_dict()
    for f in floors
}

zero_depts = set(zero["D1"]) | set(zero["D2"])
zero_blocks = blocks[blocks["Department_Sub-Department"].isin(zero_depts)]
zero_blocks = zero_blocks.sort_values("Area", ascending=False)

for idx, row in zero_blocks.iterrows():
    d, a = row["Department_Sub-Department"], row["Area"]
    best_f, best_dev = None, 1e9

    for f in floors:
        free = floor_capacity[f] - used_area[f]
        if a > free:
            continue
        cur = dept_used[f].get(d, 0) / floor_capacity[f]
        dev = abs(cur - global_ratio[d])
        if dev < best_dev:
            best_dev, best_f = dev, f

    if best_f is None:
        best_f = max(floors, key=lambda f: floor_capacity[f] - used_area[f])

    blocks.at[idx, "floor"] = best_f
    used_area[best_f] += a
    dept_used[best_f][d] = dept_used[best_f].get(d, 0) + a

# ——— 8) EXPORT FINAL ASSIGNMENTS —————————————————————
out = blocks[[
    "Block ID",
    "Department_Sub-Department",
    "Area",
    "floor"
]].rename(columns={
    "Department_Sub-Department": "SubDept",
    "floor":                    "AssignedFloor"
})

out.to_excel(
    "/content/final_floor_assignments1.xlsx",
    index=False,
    sheet_name="Assignments"
)

print("✅ Done → /content/final_floor_assignments1.xlsx")
```
